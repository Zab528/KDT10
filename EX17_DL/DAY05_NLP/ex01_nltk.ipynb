#### 【 자연어 처리용 형태소분석 - NLTK 】

```python
## 모듈 로딩
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.tokenize import PunktSentenceTokenizer, WordPunctTokenizer
```

[형태소 분석 - 단어/문장 단위 분리] <hr>

```python
## -------------------------------------------------------
## [1] 단어 단위 형태소 분석
## -------------------------------------------------------
text="Happy, New Year! Don't stop."

## -> 분리 규칙 : 구두점과 문장 토큰화 기반
tkList = word_tokenize(text)

## -> 토큰화 인스턴스 생성 및 메서드 사용
psTK    = PunktSentenceTokenizer()
tkList2 = psTK.tokenize(text)

print("[tkList]  =>", tkList)
print("[tkList2] =>", tkList2)
```

```python
## -------------------------------------------------------
## [2] 문장 단위 형태소 분석
## -------------------------------------------------------
text = 'The Matrix is everywhere its all around us, here even in this room. You can see it out your window or on your television. You feel it when you go to work, or go to church or pay your taxes.'

sent_tokenize(text)
```

```python
## -------------------------------------------------------
## [3] 텍스트 형태소 분석 
## -------------------------------------------------------
text = 'The Matrix is everywhere its all around us, here even in this room. You can see it out your window or on your television. You feel it when you go to work, or go to church or pay your taxes.'

## 영어 경우 대소문자 일치 => 일반적 소문자
text = text.lower()

## 텍스트의 모든 토큰 저장
all_tk  = set()     ## 자동 중복 제거
sent_tk = []

## 텍스트 => 문장 분리
sentList = sent_tokenize(text)

## 문장 => 단어 토큰 분리
for sent in sentList:
    word_tk = word_tokenize(sent)
    sent_tk.append(word_tk)
    all_tk.update(word_tk)      ## 여러개 요소 추가 위해서
```

```python
## 확인
for tk in sent_tk: print(tk) 

print(f'모든 토큰들 : {len(all_tk)}개\n', all_tk)
```
