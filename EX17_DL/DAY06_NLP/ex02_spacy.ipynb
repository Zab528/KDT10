```python
## 모듈로딩
import spacy
```

```python
## 한국어 형태소 분석 모델 로딩 및 인스턴스 생성
KO_MODEL = 'ko_core_news_sm'

koModel = spacy.load(KO_MODEL)
```

```python
## 형태소 분석 진행
text = "일만 하고 놀지 않으면 바보가 된다."

ret = koModel(text)
print(f"ret => {type(ret)} : {ret}")
```

```python
## token 인스턴스의 속성 확인
for token in ret:
    print(f'text 속성 : {token.text:<5} 품사 : {token.pos_:<8}', end='')
    print(f'상세 : {token.tag:<5} 어근 : {token.lemma_:<5}')
```

```python
## 품사가 명사인 토큰만 출력
for token in ret:
    if token.pos_ == 'NOUN': print(token.text)
```
