### 【 Model save & load 】
- 2가지 형태 저장
    * 형태 저장
    * 모델의 파라미터만 저장
- 2가지 형태 로딩
    * 전체 저장 모델 파일 ==> 로딩으로 사용 가능
    * 모델 파라미터 저장 파일 ==> 모델 인스턴스 생성 후 층별 파라미터 적용

[1] 모듈 로딩 및 데이터 준비 <hr>

```python
## [1-1] 모듈 로딩
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
```

```python
## [1-2] 데이터 준비
featureTS = torch.tensor([[1,5], [2,6], [3,7], [4,8], [5,9]], dtype=torch.float32)
targetTS  = torch.tensor([[6], [8], [10], [12], [14]], dtype=torch.float32)
```

[2] 모델 클래스 정의 <hr>

```python
class Test(nn.Module):
    def __init__(self):
        super().__init__()
        self.fc1 = nn.Linear(2, 5)
        self.fc2 = nn.Linear(5, 7)
        self.out = nn.Linear(7, 1)

    def forward(self, x):
        out = F.relu(self.fc1(x))
        out = F.relu(self.fc2(out))
        return self.out(out)
```

[3] 학습 진행 <hr>

```python
## 설정값들
EPOCHS = 10
BS = 2
LR = 0.1

## 저장 모델 파일명
ALL_MODEL = './all_model.pt'    ## 모델 전체 확장자    pt
WEIGHTS_MODEL = './weights'     ## 파라미터 저장 확장자 pth
```

```python
## 인스턴스들
model       = Test()
loss_fn     = nn.MSELoss()
optimizer   = optim.Adam(model.parameters(), lr=LR)
```
