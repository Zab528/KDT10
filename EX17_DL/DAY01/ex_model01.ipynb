### 【 ANN 모델 구성 】

- Linear Layer 클래스

    - 입력값으로 피쳐 수
    - 출력값으로 출력 수 ==> 층의 퍼셉트론 개수
- nn.Module 클래스
    - 커스텀 모델 구현을 위한 부모 클래스
