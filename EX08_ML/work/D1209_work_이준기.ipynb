### 인코딩 함수 만들기
- 함수 종류
    * 범주형 (명목형 / 순서형)
    * 수치형

## 범주형
- 명목형
    * OneHotEncoder

```python
def encode_onehot(df, cols):

    from sklearn.preprocessing import OneHotEncoder
    import pandas as pd

    ohEncoder = OneHotEncoder(sparse_output=False)

    ohEncoder.fit(df[cols])

    encoding = ohEncoder.transform(df[cols])

    return encoding
```

- 순서형
    * OrdinalEncoder

```python
def encode_ordi(df, cols):

    from sklearn.preprocessing import OrdinalEncoder
    import pandas as pd

    ordEncoder = OrdinalEncoder()

    ordEncoder.fit(df[cols])

    encoding = ordEncoder.transform(df[cols])

    return encoding
```

## 수치형
- 정규 스케일링

```python
def scaling(df, cols):
    from sklearn.preprocessing import StandardScaler

    scaler = StandardScaler()

    scaled_data = scaler.fit_transform(df[cols])

    return scaled_data
```
