[1] 모듈 로딩 및 데이터 준비 <hr>

```python
## 모듈 로딩
import cv2
import numpy as np
import matplotlib.pyplot as plt
import koreanize_matplotlib
```

```python
## 데이터 준비 : 도화지랑 이미지
IMG_FILE = '../Data/Image/lion_face.jpg'
PAPER_FILE = '../Data/Image/paper_w.jpg'
```

```python
## 흰색 도화지 이미지 생성 및 저장
ret=cv2.imwrite(PAPER_FILE,
            np.full((500, 500), 255, dtype=np.uint8))
print(f'save ok? {ret}')
```

```python
## 흰색 도화지 이미지 데이터 & 사진 이미지 데이터 로딩
paper = cv2.imread(PAPER_FILE)
picture = cv2.imread(IMG_FILE)

print(f'[paper]   {paper.shape}, {paper.ndim}D차원')
print(f'[picture] {picture.shape}, {picture.ndim}D차원')
```

[2] 그리기 관련 기능들 <hr>

```python
## [2-1] 직선 그리기
cv2.line(paper, (50,50), (150,50), (255,0,0))
cv2.imshow('line', paper)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
## [2-2] 사각형 그리기
cv2.rectangle(paper, (450,200), (200,450), (0,0,255), -1)
cv2.imshow('rectangle', paper)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
## [2-3] 다각형 그리기
pts4 = np.array([[350,250], [450,350], [400,450], [300,450], [250,350]], dtype=np.int32)
cv2.polylines(paper, [pts4], True, (0,0,0))
cv2.imshow('polylines', paper)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
## [2-4] 원 그리기
cv2.circle(paper, (300, 150), 70, (0,255,0), 5)
cv2.imshow('circle', paper)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
## [2-5] 글자 쓰기
cv2.putText(paper, "Maple", (200, 110), cv2.FONT_HERSHEY_SIMPLEX, 2, (0,0,250))
cv2.imshow('draw text', paper)
cv2.waitKey()
cv2.destroyAllWindows()
```
