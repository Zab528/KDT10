[1] 모듈 로딩 및 데이터 준비 <hr>

```python
## 모듈 로딩
import cv2
import numpy as np
import matplotlib.pyplot as plt
import koreanize_matplotlib
```

```python
## 데이터 준비 : 도화지랑 이미지
IMG_FILE = '../Data/Image/lion_face.jpg'
PAPER_FILE = '../Data/Image/paper_w.jpg'
```

```python
## 흰색 도화지 이미지 생성 및 저장
ret=cv2.imwrite(PAPER_FILE,
            np.full((500, 500), 255, dtype=np.uint8))
print(f'save ok? {ret}')
```

```python
## 흰색 도화지 이미지 데이터 & 사진 이미지 데이터 로딩
paper = cv2.imread(PAPER_FILE)
picture = cv2.imread(IMG_FILE)

print(f'[paper]   {paper.shape}, {paper.ndim}D차원')
print(f'[picture] {picture.shape}, {picture.ndim}D차원')
```
