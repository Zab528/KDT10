#### 【 OpenCV - 이미지 저장 】

- 함수: cv2.imwrite('이미지파일경로', 이미지로우데이터)
- 결과: 저장 여부
- ★★★ 주의 : opencv 한글 미지원! 경로, 파일명 한글 사용 XXX

[1] 모듈 로딩 및 데이터 준비<hr>

```python
## 이미지 처리 모듈
import cv2
## 시각화 모듈
import matplotlib.pyplot as plt
import koreanize_matplotlib
```

```python
## 이미지 파일
IMG_FILE = '../Data/Image/cat_02.jpg'
```

[2] 이미지 로딩 및 정보 확인 <hr>

```python
## 원본 & 회색 로딩
org_img = cv2.imread(IMG_FILE)
gray_img = cv2.imread(IMG_FILE, cv2.IMREAD_GRAYSCALE)
```

```python
## 이미지 데이터 정보 확인
print(f'[Original IMAGE]')
print(f'형태 : {org_img.shape}  차원 : {org_img.ndim}D  사이즈 : {org_img.size}')

print(f'[Gray IMAGE]')
print(f'형태 : {gray_img.shape}  차원 : {gray_img.ndim}D  사이즈 : {gray_img.size}')
```

```python
## 이미지 시각화
fig, axes = plt.subplots(1,2, sharey=True)

axes[0].imshow(org_img)
axes[0].set_title("컬러이미지")
axes[0].set_xticklabels([])
axes[0].set_yticklabels([])

axes[1].imshow(gray_img, cmap='gray')
axes[1].set_title("흑백이미지")
axes[1].set_xticklabels([])
axes[1].set_yticklabels([])

plt.suptitle("이미지 시각화")
plt.show()
```

[3] 이미지 저장 <hr>

```python
## 저장 이미지 경로+파일명
SAVE_FILE = './gray_cat.jpg'

ret = cv2.imwrite(SAVE_FILE, gray_img)

print(f'저장결과 : {ret}')
```

```python
## ==============================================================================
## OpenCV는 이미지 경로, 창 제목 등 한글 미지원!!! 동작하지 않음!!! (x키 누르면 동작 정지)
## ==============================================================================
while True:
    if org_img is not None:
        cv2.imshow("ORG", org_img)              ## 이미지 창 준비 및 띄우기

        key=cv2.waitKey()                       ## 이미지 창에 사용자 이벤트 대기

        if key in [ord('S'), ord('s')]:
            ret = cv2.imwrite('./copy.jpg', org_img)
            print(f"save ok!") if ret else print(f"save X!!")

        elif key in [ord('X'), ord('x')]:
            cv2.destroyAllWindows()             ## 이미지 창 닫기
            break
    else:
        print("None image")
        break
```

```python
## ============================================================
## 일반이미지
## ============================================================
if org_img is not None:
    print("image ok")
    cv2.imshow("ORG", org_img)          ## 이미지 창 준비 및 띄우기
    cv2.waitKey(2000)                   ## 2000ms(2sec) 이벤트 대기
    cv2.destroyAllWindows()             ## 이미지 창 닫기
else:
    print("None image")
```
