#### 【 OpenCV - 이미지 로딩 및 픽셀 데이터 생성 】

- 함수: cv2.imread('이미지파일경로', 읽기_플래그)
- 결과: ndarray 객체
- 기본: color 이미지
- 특징: 읽을 때 읽기_플래그 설정에 따라서 변환 후 로딩 가능
- ★★★ 주의 : opencv의 채널 순서가 BGR로 일반 이미지 순서와 다름

[1] 모듈 로딩 및 데이터 준비<hr>

```python
## 이미지 처리 모듈
import cv2
## 시각화 모듈
import matplotlib.pyplot as plt
import koreanize_matplotlib
```

```python
## 이미지 파일
IMG_FILE = '../Data/Image/flower_rainbow.jpg'
```

[2] 이미지 로딩 및 정보 확인 <hr>

```python
## 원본 & 회색 로딩
org_img = cv2.imread(IMG_FILE)
gray_img = cv2.imread(IMG_FILE, cv2.IMREAD_GRAYSCALE)
```

```python
## 이미지 데이터 정보 확인
print(f'[Original IMAGE]')
print(f'형태 : {org_img.shape}  차원 : {org_img.ndim}D  사이즈 : {org_img.size}')

print(f'[Gray IMAGE]')
print(f'형태 : {gray_img.shape}  차원 : {gray_img.ndim}D  사이즈 : {gray_img.size}')
```

```python
## 이미지 시각화
fig, axes = plt.subplots(1,2, sharey=True)

axes[0].imshow(org_img)
axes[0].set_title("컬러이미지")
axes[0].set_xticklabels([])
axes[0].set_yticklabels([])

axes[1].imshow(gray_img, cmap='gray')
axes[1].set_title("흑백이미지")
axes[1].set_xticklabels([])
axes[1].set_yticklabels([])

plt.suptitle("이미지 시각화")
plt.show()
```

[3] 이미지 출력 및 이벤트 대기 <hr>

- GUI 프로그램은 사용자 입력에 따라 프로그램 구동
- 사용자의 키이벤트에 따라 이미지 출력창 제어
- 단!!! 반드시 이미지 출력창에 포커스

```python
## ============================================================
## OpenCV는 이미지 경로, 창 제목 등 한글 미지원!!! 동작하지 않음!!!
## ============================================================
while True:
    if org_img is not None:
        print("image ok")
        cv2.imshow("ORG", org_img)          ## 이미지 창 준비 및 띄우기
        cv2.waitKey()                       ## 이미지 창에 사용자 이벤트 대기
        cv2.destroyAllWindows()             ## 이미지 창 닫기
        break
    else:
        print("None image")
        break
```

```python
## ============================================================
## 일반이미지
## ============================================================
if org_img is not None:
    print("image ok")
    cv2.imshow("ORG", org_img)          ## 이미지 창 준비 및 띄우기
    cv2.waitKey(2000)                   ## 2000ms(2sec) 이벤트 대기
    cv2.destroyAllWindows()             ## 이미지 창 닫기
else:
    print("None image")
```
