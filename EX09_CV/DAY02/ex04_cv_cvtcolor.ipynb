#### 【 OpenCV - 이미지 색공간(Color Space) 변경 】

- 목적에 따른 색상 표현해주는 색공간 변경 위한 것
    - 함수: cv2.cvtColor(로우이미지데이터, code)
        * code 매개변수 : cv2.COLOR_BGR2RGB, cv2.COLOR_BGR, ....
    - 결과: 새로운 색공간 적용된 이미지 로우 데이터
    
- ★★★ 주의 : opencv 한글 미지원! 경로, 파일명 한글 사용 XXX

[1] 모듈 로딩 및 데이터 준비<hr>

```python
## 이미지 처리 모듈
import cv2

## 시각화 모듈
import matplotlib.pyplot as plt
import koreanize_matplotlib
```

```python
## 이미지 파일
IMG_FILE = '../Data/Image/cat_02.jpg'
```

[2] 이미지 로딩 및 정보 확인 <hr>

```python
## 원본 & 회색 로딩
org_img = cv2.imread(IMG_FILE)
gray_img = cv2.imread(IMG_FILE, cv2.IMREAD_GRAYSCALE)
```

```python
## 이미지 데이터 정보 확인
print(f'[Original IMAGE]')
print(f'형태 : {org_img.shape}  차원 : {org_img.ndim}D  사이즈 : {org_img.size}')

print(f'[Gray IMAGE]')
print(f'형태 : {gray_img.shape}  차원 : {gray_img.ndim}D  사이즈 : {gray_img.size}')
```

```python
## 이미지 시각화
fig, axes = plt.subplots(1,2, sharey=True)

axes[0].imshow(org_img)
axes[0].set_title("컬러이미지")
axes[0].set_xticklabels([])
axes[0].set_yticklabels([])

axes[1].imshow(gray_img, cmap='gray')
axes[1].set_title("흑백이미지")
axes[1].set_xticklabels([])
axes[1].set_yticklabels([])

plt.suptitle("이미지 시각화")
plt.show()
```

[3] 이미지 색 공간 변경 비교 <HR>

```python
## =============================================
## [3-1] BGR=>RGB 색공간
## =============================================
## - packing 방식
channels = cv2.split( org_img )
print(f'channels   : {len(channels)}개 {type(channels[0])}')

## - unpacking 방식
b, g, r = cv2.split( org_img )
print(f'b channels : {type(b)}, {b.shape}')
print(f'g channels : {type(g)}, {g.shape}')
print(f'r channels : {type(r)}, {r.shape}')
```

```python
## =============================================
## [3-2] BGR=>HSV 색공간
## =============================================
hsv_img = cv2.cvtColor(org_img, cv2.COLOR_BGR2HSV)
print(f'hsv_img  : {hsv_img.shape}, {hsv_img.ndim}D')

hsvf_img
```

```python
## =============================================
## [3-2] BGR=>HSV 색공간
## =============================================
hsv_img = cv2.cvtColor(org_img, cv2.COLOR_BGR2HSV)
print(f'hsv_img  : {hsv_img.shape}, {hsv_img.ndim}D')
```

```python
## - 채널별 이미지 출력
cv2.imshow("Blue Channel", b)
cv2.imshow("Green Channel", g)
cv2.imshow("Red Channel", r)

## - 무한대기 중 어떤 키든 입력 시 모든 창 닫기
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
## =============================================
## [3-2] ndarray로 채널 분리
## =============================================
print(f'org_img  shape => {org_img.shape}')  # (높이/행, 너비/열, 채널)

## 축변경 (높이/행, 너비/열, 채널) -> (채널, 행, 열)
## ndarray.shape(321, 380, 3)  ->   축0, 축1, 축2
org_img2=org_img.transpose(2, 0, 1)
print(f'org_img2.shape => {org_img2.shape}')
```

```python
## b, g, r 채널 분리
blueArray  = org_img2[0, :, :]
greenArray = org_img2[1]
redArray   = org_img2[2]
```

```python
## 채널별 시각화
titles=['RGB', 'Blue', 'Green', 'Red']
rawImg=[org_img, org_img2[0], org_img2[1], org_img2[2]]

fig, axes = plt.subplots(1, 4, sharey=True)
for idx, ax in enumerate(axes):
    ax.imshow(rawImg[idx], cmap='gray')
    ax.set_title(f'[{titles[idx]}]')
    
plt.tight_layout()
plt.show()
```

[4] 이미지 데이터 병합 <hr>

```python
## ======================================================
## [4-1] 병합관련 함수 cv2.merge((채널이미지데이터, 채널이미지데이터,..))
## ======================================================
## RGB로 병합 (red, green, blue)
#rgb_img = cv2.merge((org_img2[2], org_img2[1], org_img2[0]))
rgb_img  = cv2.merge((r, g, b))
```

```python
## ======================================================
## [4-2] BGR이미지와 RGB이미지 시각화
## ======================================================
fig,(ax1, ax2) = plt.subplots(1, 2, sharey=True)

ax1.imshow(org_img)
ax1.set_title('CV BGR IMAGE')

ax2.imshow(rgb_img)
ax2.set_title(f'RGB IMAGE')

plt.tight_layout()
plt.show()
```
