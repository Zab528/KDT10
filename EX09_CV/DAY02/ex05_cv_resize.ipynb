#### 【 OpenCV - 이미지 크기 변경 】

- 이미지의 확대/축소 기능
    - 함수: cv2.resize(로우이미지데이터, dsSize=(절대w, 절대h), fx=상대크기, fy=상대크기, interpolation=보간법)
    - 결과: 변경된 크기의 이미지 로우 데이터

[1] 모듈 로딩 및 데이터 준비<hr>

```python
## 이미지 처리 모듈
import cv2

## 시각화 모듈
import matplotlib.pyplot as plt
import koreanize_matplotlib
```

```python
## 이미지 파일
IMG_FILE = '../Data/Image/cat_02.jpg'
```

[2] 이미지 로딩 및 정보 확인 <hr>

```python
## 원본 & 회색 로딩
org_img = cv2.imread(IMG_FILE)
gray_img = cv2.imread(IMG_FILE, cv2.IMREAD_GRAYSCALE)
```

```python
## 이미지 데이터 정보 확인
print(f'[Original IMAGE]')
print(f'형태 : {org_img.shape}  차원 : {org_img.ndim}D  사이즈 : {org_img.size}')

print(f'[Gray IMAGE]')
print(f'형태 : {gray_img.shape}  차원 : {gray_img.ndim}D  사이즈 : {gray_img.size}')
```

```python
## 이미지 시각화
fig, axes = plt.subplots(1,2, sharey=True)

axes[0].imshow(org_img)
axes[0].set_title("컬러이미지")
axes[0].set_xticklabels([])
axes[0].set_yticklabels([])

axes[1].imshow(gray_img, cmap='gray')
axes[1].set_title("흑백이미지")
axes[1].set_xticklabels([])
axes[1].set_yticklabels([])

plt.suptitle("이미지 시각화")
plt.show()
```

[3] 이미지 크기 <HR>

```python
## =============================================
## [3-1] 절대 크기 변경 : 축소, 원본, 확대
## =============================================
##-> 이미지 축소 : 현재의 1/3
down_img = cv2.resize(IMG_FILE, dsize=(org_img.shape[0]/3, org_img.shape[1]/3))

##-> 이미지 확대 : 현재의 2배
up_img = cv2.resize(IMG_FILE, dsize=(0, 0), fx=1.5, fy=1.5, interpolation=cv2.INTER_CUBIC)

##-> 축소, 원본, 확대 이미지 시각화
fig, axes = plt.subplots(1, 3, sharey=True)
axes[0].imshow(down_img)
axes[1].imshow(org_img)
axes[2].imshow(up_img)
```

[4] 이미지 데이터 병합 <hr>

```python
## ======================================================
## [4-2] BGR이미지와 RGB이미지 시각화
## ======================================================
fig,(ax1, ax2) = plt.subplots(1, 2, sharey=True)

ax1.imshow(org_img)
ax1.set_title('CV BGR IMAGE')

ax2.imshow(rgb_img)
ax2.set_title(f'RGB IMAGE')

plt.tight_layout()
plt.show()
```
