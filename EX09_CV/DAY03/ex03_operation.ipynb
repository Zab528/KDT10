#### 【 이미지 연산 】

- 픽셀(Pixel) 연산 수행 이미지 변화
    * 이미지데이터 => 행열, Numpy의 ndarray
    * 덧__셈 연산 => 각각의 픽셀 단위 덧셈
    * 뺄__셈 연산 => 각각의 픽셀 단위 뺄셈
    * 곱__셈 연산 => 각각의 픽셀 단위 곱셈
    * 나눗셈 연산 => 각각의 픽셀 단위 나눗셈
- 색상값 : 0 검정색 Black - 1 또는 255 흰색 White
    * 0 >>>>>>> 1 또는 255 : 밝아짐
    * 0 <<<<<<< 1 또는 255 : 어두워짐
- 이미지 밝기 조절
- 이미지 자연스러운 합성

[1] 모듈 로딩 및 데이터 준비 <hr>

```python
# 모듈 로딩
import numpy as np
import cv2
import matplotlib.pyplot as plt
import koreanize_matplotlib

# 데이터 파일
IMG_FILE = '../Data/Image/flower.jpg'
```

```python
## 데이터 로딩 및 기본 정보 출력
imgNP = cv2.imread(IMG_FILE)
grayNP = cv2.imread(IMG_FILE, cv2.IMREAD_GRAYSCALE)

print(f'imgNP: {imgNP.shape}, grayNP: {grayNP.shape}') 
```

[2] 데이터 로딩 <hr>

```python
## 원본 이미지 데이터 로딩
oriNP  = cv2.imread(IMG_FILE)
grayNP = cv2.imread(IMG_FILE, cv2.IMREAD_GRAYSCALE)
```

```python
## 이미지 기본 정보 확인
print(f'oriNP => {oriNP.shape} grayNP => {grayNP.shape}')
```

[3] 이미지 밝기 조절 <hr>

```python
## [3-1] 컬러 이미지 밝기 조절
### 0 >>>>> 255 : 흰색, White 밝아짐
colorNP30  = cv2.add(oriNP, 30)
colorNP50  = cv2.add(oriNP, 50)
colorNP100 = cv2.add(oriNP, 100)
```

```python
### 시각화
fig, axes   = plt.subplots(1, 4, figsize=(20,5))
numObj      = [oriNP, colorNP30, colorNP50, colorNP100]
titles      = ['ORI', '+30', '+50', '+100']

for ax, img, title in zip(axes, numObj, titles):
    ax.imshow(img)
    ax.set_title(title)
    ax.set_axis_off()
plt.show()
```

```python
### 0 <<<<< 255 : 검은색, Black 어두워짐
colorNP30  = cv2.subtract(oriNP, 30)
colorNP50  = cv2.subtract(oriNP, 50)
colorNP100 = cv2.subtract(oriNP, 100)
```

```python
### 시각화
fig, axes   = plt.subplots(1, 4, figsize=(20,5))
numObj      = [oriNP, colorNP30, colorNP50, colorNP100]
titles      = ['ORI', '-30', '-50', '-100']

for ax, img, title in zip(axes, numObj, titles):
    ax.imshow(img)
    ax.set_title(title)
    ax.set_axis_off()
plt.show()
```
