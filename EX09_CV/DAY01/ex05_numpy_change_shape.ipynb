#### 【 Numpy - ndarray 형태 변경 】

- 1차원 => N차원, N차원 => 1차원 : 원소 개수 변경 안됨!!!

- 1차원 <=> N차원
    + np.reshape()함수, ndarray.reshape()메서드
    + np.resize()함수, ndarray.resize()메서드 **<== 예외**

- N차원 => 1차원
    + np.flatten()함수, ndarray.flatten() 메서드
    + np.ravel()함수, ndarray.ravel()메서드

```python
## 모듈 로딩
import numpy as np          ## 이미지 데이터 저장 관련 모듈
```

- 임의의 데이터 생성 및 형변환 <hr>

```python
## 임의의 데이터 ndarray 객체 생성
a = np.arange(11, 23)
print(a.shape, a.dtype, a, sep='\n')
```

```python
## =================================================
## 다양한 형태(shape) 변경
## =================================================
## 1D (12,) => 2D : (1, 12), (2, 6), (3, 4), (4, 3), (6, 2), (12, 1)
d2 = a.reshape((3,4))
print('(3, 4) 변경 후', d2.shape)

d2 = a.reshape((3,-1))
print('(3,-1) 변경 후', d2.shape)

d2 = a.reshape((-1, 1))
print('(-1, 1) 변경 후', d2.shape)
```

```python
## 1D (12,) => 3D : (1, 1, 12), (2, 1, 6), (6, 2, 1)
d2 = a.reshape((1, 1, 12))
print('(1, 1, 12) 변경 후', d2.shape, d2.ndim)

d2 = a.reshape((-1, 2, 1))
print('(-1, 2, 1) 변경 후', d2.shape, d2.ndim)

d2 = a.reshape((2, -1, 2))
print('(2, -1, 2) 변경 후', d2.shape, d2.ndim)

d2 = np.reshape(a, (2, -1, 2))
print('(2, -1, 2) 변경 후', d2.shape, d2.ndim)


d2 = np.reshape(a, (-1,))
print('(-1,) 변경 후', d2.shape, d2.ndim)
```

```python
## 원소 개수 변경 : 12개 ---> 18개 ValueError 발생

#d2 = a.reshape((3, 3, 2))   ## 3 * 3 * 2 => 18개 원소
print('(3, 3, 2) 변경 후', d2.shape, d2.ndim)

#d2 = np.reshape(a, (3, 3, 2))
print('(3, 3, 2) 변경 후', d2.shape, d2.ndim)
```

```python
## =================================================
## 다양한 형태(shape) 변경 / 원소 개수 변경 가능
## =================================================
## 1D (12,) => 2D : (1, 12), (2, 6), (3, 4), (4, 3), (6, 2), (12, 1)
##-> 원본 바로 적용
a.resize((3,4))
print('(3, 4) 변경 후', a)

##-> -1 사용 불가 X
# a.resize((-1, 6))
# print('(-1, 6) 변경 후', a)

##-> 함수 사용
d3=np.resize(a, (1, 2, 6))
print('(1, 2, 6) 변경 후', d3)


##-> 함수 사용 + 원소 개수 변경
d3=np.resize(a, (1, 6))
print('(1, 6) 변경 후', d3)

d3=np.resize(d3, (1, 8))
print('(1, 8) 변경 후', d3)


##=> 메서드 적용 + 원소 개수 변경
a = np.arange(11, 23)
print(a.shape, a.dtype, a, sep='\n')

a.resize((2,8), refcheck=False)
print(a.shape, a.dtype, a, sep='\n')
```

```python
## =================================================
## 다양한 형태(shape) 변경 : N차원 ---> 1차원
## =================================================
## 방법1) reshape(), resize()
## -> reshape() : 원본 적용 X, -1 사용 가능

a = np.array([[11, 12], [13, 14], [15, 16]])
print(a.shape, a, sep='\n')

a1 = a.reshape(-1)
print(a.shape, a1, sep='\n')

## 방법1) reshape(), resize()
## -> resize() : 원본 적용 O, -1 사용 불가능
a.resize(6)
print(a.shape, a, sep='\n')
```

```python
## 방법2) flatten()
## -> flatten() : 복사본 생성 후 1차원 변환. 속도가 상대적으로 느림
a = np.array([[[11, 12], [13, 14], [15, 16]]])
print("원본 =>", a.shape, a, sep='\n')

a1 = a.flatten()
print("1차원 변형 => ", a1)

a[0][0][0] = 77
print("원본 a  => ", a)
print("변형 a1 => ", a1)
```

```python
## 방법3) ravel()
## -> ravel() : 원본 공유해서 1차원 변환.
a = np.array([[[11, 12], [13, 14], [15, 16]]])
print("원본 =>", a.shape, a, sep='\n')

a1 = a.ravel()
print("1차원 변형 => ", a1)

a[0][0][0] = 77
print("원본 a  => ", a)
print("변형 a1 => ", a1)
```
