#### 【 Numpy - ndarray 형태 변경 】

- 1차원 => N차원, N차원 => 1차원 : 원소 개수 변경 안됨!!!

- 1차원 <=> N차원
    + np.reshape()함수, ndarray.reshape()메서드
    + np.resize()함수, ndarray.resize()메서드

- N차원 => 1차원
    + np.flatten()함수, ndarray.flatten() 메서드
    + np.ravel()함수, ndarray.ravel()메서드

```python
## 모듈 로딩
import numpy as np          ## 이미지 데이터 저장 관련 모듈
```

- 임의의 데이터 생성 및 형변환 <hr>

```python
## 임의의 데이터 ndarray 객체 생성
a = np.arange(11, 23)
print(a.shape, a.dtype, a, sep='\n')
```

```python
## 방법 1) astype() 메서드 활용
a = a.astype('float16')

print('변경 후', a, a.dtype)
```

```python
## 방법 2) np.타입이름XX(ndarray) 함수 활용
a = np.uint8(a)

print('변경 후', a, a.dtype)
```
