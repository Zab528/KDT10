#### 【 Numpy - 범위내 데이터 활용 ndarray 생성 】

- numpy.random.rand() / randn() : 표준정규 분포 (평균:0, 분산:1) 기반 난수 생성
- numpy.arange() : 지정된 범위 내의 정수 배열 생성
- numpy.linspace() : 지정된 범위 내의 균일한 간격의 수치 배열 생성
- numpy.logspace() : 지정된 범위 내의 로그스케일 기반 수치 배열 생성

```python
## 모듈 로딩
import numpy as np          ## 이미지 데이터 저장 관련 모듈
```

- 임의의 데이터 생성 및 형변환 <hr>

```python
## 임의의 데이터 ndarray 객체 생성
a = np.arange(11,21,3)
print(a.shape, a.dtype, a, sep='\n')
```

```python
## 방법 1) astype() 메서드 활용
a = a.astype('float16')

print('변경 후', a, a.dtype)
```

```python
## 방법 2) np.타입이름XX(ndarray) 함수 활용
a = np.uint8(a)

print('변경 후', a, a.dtype)
```
