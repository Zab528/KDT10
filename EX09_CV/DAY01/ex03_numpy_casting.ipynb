#### 【 Numpy - 자료형 변환/형변환/캐스팅 】

- 묵시적 형변환 : 작은 데이터 타입 => 큰 데이터 타입
    * 시스템에서 자동 진행
- 명시적 형변환 : 큰 데이터 타입 => 작은 데이터 타입
    * 개발자가 진행. 데이터 손실 발생
    * 일반적으로 형변환 의미가 명시적 형변환

```python
## 모듈 로딩
import numpy as np          ## 이미지 데이터 저장 관련 모듈
```

- 임의의 데이터 생성 및 형변환 <hr>

```python
## 임의의 데이터 ndarray 객체 생성
a = np.arange(11,21,3)
print(a.shape, a.dtype, a, sep='\n')
```

```python
## 방법 1) astype() 메서드 활용
a = a.astype('float16')

print('변경 후', a, a.dtype)
```

```python
## 방법 2) np.타입이름XX(ndarray) 함수 활용
a = np.uint8(a)

print('변경 후', a, a.dtype)
```
