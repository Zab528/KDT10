#### 【 이미지 분류 모델을 위한 준비 】

- 데이터 준비 : 수집된 데이터 ==> 하나의 데이터로 저장
- scikit-learn
    * 피쳐 데이터 : 2D ==> 이미지 파일 1개 ==> 1개 행 데이터. 픽셀이 컬럼이 됨!
    * 타겟 데이터 : 1D ==> 이미지 파일명

- 데이터 즉, 피쳐와 타겟 => csv, json, xlsx, ... 저장

```python
import cv2
import os

DATA_DIR = "../PrData/B"
CSV_DIR  = "../PrData/csv"
CSV_FILE = "../PrData/csv/B.csv"

os.makedirs(CSV_DIR, exist_ok=True)

with open(CSV_FILE, mode="w", encoding="utf-8") as f:   # 처음부터 새로 만들기
    for filename in os.listdir(DATA_DIR):
        img_path = os.path.join(DATA_DIR, filename)

        img = cv2.imread(img_path)  # 컬러(BGR)
        if img is None:
            continue

        img = cv2.resize(img, (0, 0), fx=0.5, fy=0.5, interpolation=cv2.INTER_AREA)
        img = cv2.resize(img, (70, 70), interpolation=cv2.INTER_AREA)

        #(원하면 RGB로 바꾸기)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

        flat = img.reshape(-1).astype(int)  # (70*70*3,)

        row = "black," + ",".join(map(str, flat.tolist())) + "\n"
        f.write(row)
```
