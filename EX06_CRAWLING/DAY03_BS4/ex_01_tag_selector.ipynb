### 【 스크래핑 TAG / SELECTOR 이해 】

[1] 모듈 로딩 및 데이터 준비<hr>

```python
from bs4 import BeautifulSoup           ## HTML STR ==> Object화 모듈
from urllib.request import urlopen      ## WEB으로부터 Data Loading 모듈
```

```python
## ---------------------------------------------------------------------------
## DATA 준비
## ---------------------------------------------------------------------------
HTML_FILE = '../Data/chap02.html'
HTML_URL  = 'https://finance.naver.com/'
```

```python
## ---------------------------------------------------------------------------
## HTML STR 로딩
## ---------------------------------------------------------------------------
## (1) html 파일
with open(HTML_FILE, mode='r', encoding='utf-8') as f:
    file_data = f.read()
print(file_data)

## (2) Web 페이지 bytes => str
url_data = urlopen(HTML_URL).read()
print(url_data)
```

[2] HTML FILE : HTML STR => BS Object 변환 <hr>

```python
## HTML TAG Object 진행 완료
fileBS = BeautifulSoup(file_data, 'html.parser')
```

```python
## ===================================================================
## 원하는 데이터 추출
## ===================================================================
## -> find(tag_str, attrs, **kwargs) 메서드 : 제일 첫번째 발견된 태그
## -> find_all(tag_str, attrs, **kwargs) 메서드 : 모든 태그
## -------------------------------------------------------------------
## -> select(selector_str)메서드        : Tag의 id, class 속성들 활용해서
##                                       원하는 태그 모두 추출
## -> select_one(selector_str)메서드    : Tag의 id, class 속성들 활용해서
##                                       1개 태그 추출
## ===================================================================
```

```python
## => class1's first paragraph 텍스트 추출
ret1 = fileBS.find('p', attrs={'id':'first'})
ret2 = fileBS.select_one('#first')

print(f'ret1 => {ret1}')
print(f'ret2 => {ret2}')
```

```python
## => naver 텍스트 추출
ret1 = fileBS.find('a', attrs={'class':'exteranl_link'})
ret1_text = ret1.get_text()

ret2 = fileBS.select_one('a.external_link')
ret2_text = ret2.get_text()

print(f'ret1 => {ret1_text}')
print(f'ret2 => {ret2_text}')
```

```python
## => naver 텍스트 추출
ret1 = fileBS.find('div')
print(ret1)

ret1.find_all('a')
ret1.find('a', attrs={'href':'www.naver.com'})
ret1.find('a', href = "www.naver.com")
ret1.find(string='naver')


## - selector 기반
ret3 = fileBS.select_one('div#class1 > a.external_link')
print(ret3)
```
