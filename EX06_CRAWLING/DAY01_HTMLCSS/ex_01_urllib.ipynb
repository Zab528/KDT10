```python
#from urllib import request as req                      ## - request 모듈 전체
from urllib.request import urlopen, urlretrieve         ## - request 모듈 내 함수 2개만
import os                                               ## - 파일/폴더 경로 체크
```

[2] URL에 해당하는 데이터 로딩 <hr>

```python
## =================================================================
## [2-1] urlretrieve 경우 ===> url의 데이터를 파일로 저장
## =================================================================
IMG_URL = r'https://search.pstatic.net/sunny/?src=https%3A%2F%2Fimages.rawpixel.com%2Fimage_social_landscape%2FcHJpdmF0ZS9sci9pbWFnZXMvd2Vic2l0ZS8yMDIzLTA4L3Jhd3BpeGVsX29mZmljZV8yNl9waG90b19vZl9hX2RhaXN5X2Zsb3dlcl9pc29sYXRlZF9jb2xvcl9iYWNrZ19kNjE2YzM2YS1jYjkxLTQ3MTYtOTdmNS0wZTQ0ZWY0ZTc1ZmVfMS5qcGc.jpg&type=a340'
SAVE_FILE = './img.jpg'

f_path, msg = urlretrieve(IMG_URL, SAVE_FILE)

if os.path.exists(SAVE_FILE):
    print(f'f_path => {f_path}')
    print(f'msg    => {msg}')
else:
    print('URL 체크 바랍니다. 저장 실패!')
```

```python
## =================================================================
## [2-2] urlopen 경우 ===> url의 데이터를 저장
##                        http.client.HTTPResponse 객체에 담아서 반환
## =================================================================
WEB_URL = r'https://search.naver.com/search.naver?ssc=tab.image.all&where=image&sm=tab_jum&query=flower'

#imgReq=urlopen(IMG_URL)
imgReq=urlopen(WEB_URL)

print(f'imgReq => {imgReq}')
print(f'속성 : imgReq.headers => {imgReq.headers}')
print(f'메서드 : imgReq.read() => {imgReq.read()}')
```

```python
# 이미지 로우 데이터 읽어서 저장

# - 웹에서 가져오기
resObj = imgReq = urlopen(IMG_URL)
# - 바이너리 데이터 추출
raw_data = resObj.read()

# 바이너리 모드로 파일 open & writ
with open(SAVE_FILE, mode='wb') as f:
    f.write(raw_data)
    print("저장되었습니다...!")
```
