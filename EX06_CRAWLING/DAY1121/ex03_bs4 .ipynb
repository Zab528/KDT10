```python
## HTML파일에서 일부부만 선택 str로 저장
html_str = """
    <h1> 풀꽃</h1>
    나태주 시인, 교육인<br>
출생1945년 3월 16일, 충청남도 서천 데뷔1971년 서울신문 신춘문예 시 '대숲 아래서' 등단<br>
<p>
[경력]<br>
2020.04 ~ 제43대 한국시인협회 회장<br>
2010.07 ~ 2017.06 공주문화원 원장<br>
한국시인협회 심의위원회 위원장<br>
1993 ~ 1994 충남시인협회 회장<br>
~ 2007.08 공주 장기초등학교 교장<br>
</p>
 
<p>
나태주                                                                         <br> 
1945년 출생으로 대표 시 「풀꽃」처럼 작고 여린 존재를 향한 시를 쓴다. 
1971년 [서울신문] 신춘문예에 「대숲 아래서」로 등단한 후 현재까지 40여 권의 창작시집을 포함해서 
100여 권의 책을 펴냈다. 초등학교 다닐 때의 꿈은 화가였으나 고등학교 1학년 때 예쁜 여학생을 만난 뒤로는 
꿈이 시인으로 바뀌었다. 그로부터 60년 그는 끝없이 시인을 꿈꾸며 사는 사람이다. 
</p>
<p>
1963년 공주사범학교를 졸업한 뒤 초등학교에서 43년간 교직 생활을 하다가 2007년 장기초등학교 교장으로 정년퇴임했으며
 8년 동안 공주문화원장을 거쳐 현재는 공주에서 공주풀꽃문학관을 설립·운영하고 있으며 풀꽃문학상, 해외풀꽃시인상, 공주문학상 등을 제정, 시상하고 있다. 
 풀꽃문학관에서, 서점에서, 도서관에서, 전국 방방곡곡 사람들을 만나러 다니는 게 요즘의 일상이다. 가깝고 조그마한, 손 뻗으면 충분히 닿을 수 있는 시인으로 기억되고 싶다. 
</p>
<p>
흙의문학상, 박용래문학상, 편운문학상, 한국시인협회상, 정지용문학상, 유심작품상 등을 받았고, 
충남문인협회 회장, 충남시인협회 회장, 공주문인협회 회장, 공주녹색연합 초대대표, 한국시인협회 심의위원장 등을 
역임했다.청소년 시절 그는 세 가지 소원이 있었다. 
첫째가 &nbsp;시인이 되는 것이었고, 둘째가 좋은 여자와 결혼하는 것이었고, 
셋째가 &nbsp;&nbsp;&nbsp;공주에 사는 것이었는데 오늘날 그 소원을 모두 이루었다고 생각한다.
</p>
<br>
"""
```

```python
## markup str ==> 객체화
mbs = BeautifulSoup(html_str, 'html.parser')
```

```python
## [1] 태그객체 전체 가져오기 : BS객체.태그명
print("p 태그 전체 : ", mbs.p)
print("p 태그 전체 : ", mbs.p)
```

```python
## find()메서드 활용
print("h1 텍스트 :", mbs.h1.string)  # 'hello'  (텍스트만 있어서 OK)
print("h2 텍스트 :", mbs.h2.string)  # None     (안에 <b> 태그가 있어서)

##- 일반적인 방법 : Tag객체명.text 또는 Tag객체명.get_text()
print("h2 텍스트 :", mbs.h2.text, mbs.h2.get_text())
print("a  텍스트 :", mbs.a.text,  mbs.a.get_text())
```

```python
## [3] 태그 속성 가져오기 : Tag변수명['속성명'] / Tag변수명.get('속성명')
## ** 속성 : 태그에 부여된 다양한 옵션
##          시작 태그 내에 속성 설정
##          [예] <a href="이동URL" target="띄우는 방법"> .... </a>
##               <table style='border: 2px solid red;'>

print("   a 태그의 href 속성값  :", mbs.a['href'])
print("   a 태그의 href 속성값  :", mbs.a.get('href'))
print(" div 태그의 style 속성값 :", mbs.div.get('style'))
print("   a 태그의 class 속성값 :", mbs.a['class'])
```
