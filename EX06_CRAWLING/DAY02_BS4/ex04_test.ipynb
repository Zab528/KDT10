```python
html_str = """
<body>
<div id="main-goods" role="page">
  <h1>과일과 야채</h1>
  <ol id="fr-list">
    <li class="red green" data-lo="ko">사과</li>
    <li class="purple" data-lo="us">포도</li>
    <li class="yellow" data-lo="us">레몬</li>
    <li class="yellow" data-lo="ko">오렌지</li>
  </ol>
  <ul id="ve-list">
    <li class="white green" data-lo="ko">무</li>
    <li class="red green" data-lo="us">파프리카</li>
    <li class="black" data-lo="ko">가지</li>
    <li class="black" data-lo="us">아보카도</li>
    <li class="white" data-lo="cn">연근</li>
  </ul>
</div>
</body>"""
```

```python
mbs = BeautifulSoup(html_str, 'html.parser')
```

```python
fr_ol = mbs.find('ol', id='fr-list')     
fr_lis = fr_ol.find_all('li') 

fruit_list=[]
for i in range(4):
    fruit_list.append(fr_lis[i].get_text())

print(fruit_list)
```

```python
ve_ul = mbs.find('ul', id='ve-list')
ve_lis = ve_ul.find_all('li')

ve_list = []
for i in range(5):
    ve_list.append(ve_lis[i].get_text())

print(ve_list)
```

```python
all_lis = mbs.find_all('li')

color_dict = {}   # {'red': ['사과', '파프리카'], ...}

for li in all_lis:
    name = li.get_text(strip=True)        # 이름
    classes = li.get('class')             # ['red', 'green'] 이런 식 / 없으면 None

    if classes is None:
        continue

    for c in classes:
        if c not in color_dict:
            color_dict[c] = []
        color_dict[c].append(name)

print("색상별 목록:")
for color, names in color_dict.items():
    print(color, ":", names)
```
