### 【 동적 웹 페이지 크롤링 】
- Selenium 패키지
    * 웹 페이지 실행 후 사용자 이벤트 진행, JS 실행 결과 내용 추출 패키지 중 하나
    * 설치 : conda install selenium 
    * 설정
        - 브라우저 기능 수행을 위한 브라우저 드라이버 설치 필요 
        - Selenium 버전 4.6이상부터는 자동 처리 됨

```python
## 모듈 로딩 
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options

import selenium 

print(f'selenium v.{selenium.__version__}')
```

[1] 코드기반 구글 브라우저 및 웹 페이지 띄우기 <hr>

```python
## =========================================================
## Selenium 4.6 이전 드라이버 드라이버 설정 
## =========================================================
chrome_path = "../chrome-win64/chrome.exe"
driver_path = "../chromedriver-win64/chromedriver.exe"

## - 웹드라이버 설정 
options = Options()
options.binary_location = chrome_path

## - 웹드라이버 서비스 설정, 크롬 브라우저 실행
service = Service(driver_path)
driver = webdriver.Chrome(service=service, options=options)

## - 특정 웹 페이지 띄우기 
driver.get("https://google.com")
print(driver.title) 
```

```python
## =========================================================
## Selenium 4.6 이상 드라이버 자동 설치/관리 (Selenium Manager)
## =========================================================
## - 크롬 브라우저 실행
driver = webdriver.Chrome()    

## - 페이지 열기   
driver.get("https://www.google.com")

## - 페이지 제목 출력
print(driver.title)                  
# driver.quit()  # 작업 끝나면 창 닫기
```

[2] 창이 보지 않는 Headless 모드 <hr>

```python
## 모듈 로딩
from selenium import webdriver
from selenium.webdriver.chrome.options import Options

## 브라우저 설정 
options = Options()
options.add_argument("--headless=new")  ## - 창 띄우지 않고 백그라운드 실행

## 웹 브라우저 로딩 및 지정된 웹 페이지 열기 
driver = webdriver.Chrome(options=options)
driver.get("https://www.google.com")
print(driver.title)

## 웹 브라우저 종료
driver.quit()
```
