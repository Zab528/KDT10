### 【 중심 극한 정리:CLT, Central Limit Theorem 】
- 표본을 여러 번 뽑아서 표본평균이 어떻게 분포하는지 확인 
    * 모집단에서 여러 번 표본을 뽑으면
    * 표본 평균들이 하나의 분포를 이룸
    * 분포는 정규분포에 가까워짐
    → 추론통계의 핵심!

```python
## ------------------------------------------------------
## 모듈 로딩
## ------------------------------------------------------
import numpy as np
import matplotlib.pyplot as plt

```

```python
# 모집단 생성
np.random.seed(0)
population = np.random.normal(loc=50, scale=10, size=100000)

```

```python
# 표본분포 만들기
sample_means = []
n = 30  # 표본 크기
num_samples = 1000

for n in [5, 10, 30, 50, 100]:
    for _ in range(num_samples):
        sample = np.random.choice(population, n, replace=True)
        sample_means.append(np.mean(sample))

    # 시각화
    plt.figure(figsize=(8,5))
    plt.hist(sample_means, bins=30)
    plt.title(f"Sampling Distribution of the Sample Mean (n={n})")
    plt.xlabel("Sample Mean")
    plt.ylabel("Frequency")
    plt.grid(True)
    plt.show()
```

```python
# 시각화
plt.figure(figsize=(8,5))
plt.hist(sample_means, bins=30)
plt.title("Sampling Distribution of the Sample Mean (n=30)")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```
