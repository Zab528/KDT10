#### 【 Spearman Rank Coefficient 】
- 연속형 데이터, 선형 또는 비선형 단조 강도 측정
- 단조 : 전체적으로 데이터의 증가 또는 감소
- 정규분포 X, 이상치 영향 약. 비선형도 가능

```python
## =================================================
## 모듈 로딩
## =================================================
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr, spearmanr
import koreanize_matplotlib
```

```python
## 재현성
np.random.seed(0)
```

```python
## =================================================
## 선형 단조 증가 (linear, monotonic)
## =================================================
x = np.linspace(-2, 2, 80)
y_linear = 2 * x + 1 + np.random.normal(sclae=0.5, size=len(x))

## 2변수에 대한 선형 관계성(person), 단조 관계성(spearman) 측정
pearson_linear, _ = pearsonr(x, y_linear)
spearman_linear, _ = spearmanr(x, y_linear)

## 시각화
plt.figure()
plt.scatter(x, y_linear)
plt.title(f"선형 단조 증가\nPearson={pearson_linear:.3f}, Spearman={spearman_linear:.3f}")
plt.xlabel("X")
plt.ylabel("Y (linear)")
plt.grid(True)
plt.show()

## 수치만 따로 출력
```
