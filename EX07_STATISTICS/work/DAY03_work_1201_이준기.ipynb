[1] 두 변수 X, Y가 함께 증가하면 공분산은 어떤 값을 가지는가?  
1) 양수 2) 음수 3) 0 4) 항상 1

- 답 : 1번
- 풀이 : 공분산은 두 변수가 함께 증가하거나 감소하는 정도를 나타내는데, X가 커질 때 Y도 커지므로 양의 방향으로 함께 움직여서 공분산은 양수이다.

---

[2] 다음 중 상관계수의 특징으로 옳은 것은?  
1) 단위(scale)의 영향을 크게 받는다  
2) 값의 범위는 -∞ ~ +∞ 이다  
3) 단위가 달라도 비교가 가능하다  
4) 값이 0이면 반드시 관계가 없다  

- 답 : 3번
- 풀이 : 상관계수는 단위를 제거한 표준화된 값이므로 단위가 달라도 괜찮다.

---

[3] 피어슨 상관계수는 어떤 관계를 측정하는가?  
1) 비선형 관계  
2) 선형 관계  
3) 순위 기반 관계  
4) 범주형 데이터 기반 관계

- 답 : 2번
- 풀이 : 피어슨 상관계수는 두 변수 간의 직선적 선형 관계 강도를 측정한다.

---

[4] 스피어만 상관계수는 어떤 상황에서 피어슨보다 적합한가?  
1) 이상치가 많고 비선형 단조 관계일 때  
2) 두 변수 모두 범주형일 때  
3) 두 값이 모두 binary일 때  
4) 순위가 필요 없을 때  

- 답 : 1번
- 풀이 : 스피어만 상관계수는 순위 기반 상관계수 이므로 실제 값이 아닌 순위를 이용해 관계를 측정하여 이상치가 많고 비선형 + 단조 관계일 때 유리하다.

---

[5] 켄달 τ(Tau) 순위 상관계수는 무엇을 비교하여 계산되는가?  
1) 평균과 분산  
2) 순위가 일치하는 쌍과 불일치 쌍  
3) 표준편차의 차이  
4) 범주형 빈도  

- 답 : 2번
- 풀이 : 켄달은 두 변수의 순위 간 일관성을 측정하는 상관계수로이다.

---

[6] 다음 중 Point-Biserial을 적용할 수 있는 경우는?  
1) 연속형 ↔ 연속형  
2) 순위형 ↔ 순위형  
3) 이진형 ↔ 연속형  
4) 범주형 ↔ 범주형  

- 답 : 3번
- 풀이 : 두 변수 중 하나는 이진형, 하나는 연속형일 때 사용하는 상관계수이다.

---

[7] Cramér’s V는 어떤 데이터에 사용하는가?  
1) 연속형 ↔ 연속형  
2) 범주형 ↔ 범주형  
3) 연속형 ↔ 순위형  
4) 이진형 ↔ 연속형  

- 답 : 2번
- 풀이 : 두 범주형 변수 간의 연관성의 크기를 측정하는 지표로 0 ~ 1 사이의 값을 가진다.

[8] 상관계수 r = -0.8 일 때 올바른 해석은?  
1) 매우 강한 음의 선형 관계  
2) 매우 약한 양의 관계  
3) 관계 없음  
4) 비선형 관계  

- 답 : 1번
- 풀이 : -1~1중 -0.8로 절댓값이 크며 음수이므로 매우 강한 음의 선형 관계이다.

[9] 공분산이 0이면 무엇을 의미하는가?  
1) X와 Y는 독립이다  
2) 선형 관계가 없다  
3) 항상 상관계수도 0이다  
4) 두 변수의 관계를 알 수 없다  

- 답 : 2번
- 풀이 : 공분산은 선형적 방향성만을 측정한다.

[10] 상관관계가 높더라도, 반드시 인과관계가 있다고 말할 수 없다.  
1) 참 2) 거짓  

- 답 : 1번
- 풀이 : 두 변수가 함께 움직이더라도 둘 중 하나가 다른 하나의 원인이 된다고는 볼 수 없다.

3) {홀수, 짝수}  
4) {성공, 실패}

[12] “주사위를 던져 짝수가 나오는 경우”는 어떤 개념인가?

[13] 동전 2번 던져 “앞면의 개수”는 어떤 확률변수인가?  
1) 이산형  
2) 연속형  
3) 범주형  
4) 정규변수  

- 답 : 1번

[14] 정규분포 특징으로 옳은 것은?  
1) 반드시 오른쪽으로 치우친 분포  
2) 평균과 중앙값이 다르다  
3) 평균을 중심으로 좌우대칭이다  
4) 이산형 분포이다  

- 답 : 3번
- 풀이 : 정규분포는 평균(뮤)를 중심으로 좌우가 완벽하게 대칭이며 평균값=중앙값=최빈값인 종 모양의 연속형 확률분포이다.

[15] Z = (X – μ) / σ 의 목적은?  
1) 단위를 없애기 위해  
2) 새로운 평균과 분산을 만들기 위해  
3) 모든 정규분포를 동일한 기준으로 비교하기 위해  
4) 모두 맞음  

- 답 : 4번
- 풀이 : Z-score는 특정 값을 표준화하는 과정으로 위 3가지의 모든 목적이 맞다.

[16] 중심극한정리에 따르면, 표본평균의 분포는 어떻게 되는가?  
1) 항상 이항분포  
2) 표본수가 크면 정규분포에 가까워진다  
3) 반드시 균일분포

- 답 : 2번
