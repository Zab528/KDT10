```python
## ===============================================================
## [문제 1] 개념 비교
## ===============================================================
## 일반 평균, 절사 평균, 가중 평균의 공통점과 차이점을 간단히 서술하세요.
## 다음 상황에서 어떤 평균을 쓰는 것이 적절한지 고르고 이유를 쓰세요.

# 일반 평균 : 전체 값을 다 더해서 개수로 나눈 평균
# 절삭 평균 : 양극단값 제거한 데이터의 합계/개수
# 가중 평균 : (가중치 * 데이터 합계)/(가중치합계). 데이터수가 적을 때

# 공통점 : 데이터의 합계를 개수로 나눈 평균을 나타낸다
# 차이점 :   1. 일반 평균은 전체 값을 다 더한다
#           2. 절삭 평균은 양극단값을 제거한다
#           3. 가중 평균은 가중치를 곱하고, 그 가중치의 합계로 나눈다. (데이터 수가 적을 때 가능하다.)

## (1) 극단값이 포함된 연봉 데이터의 대표값
#      - 절삭 평균 사용
# 이유 : 연봉은 직급 별로 차이가 크게 나고, CEO 같이 높은 사람들의 연봉은
#        일반 사원과 차이가 많이 나기 때문에 이런 극단값들을 자르고 평균을 내는 것이 옳다.

## (2) 각기 다른 표본 수를 가진 실험 그룹들의 평균 반응값 통합
#      - 가중 평균
# 이유 : 실험마다 표본 수가 다르기 때문에, 표본 수가 많은 쪽에 더 큰 가중치를 주어서
#        더 높은 영향력을 줄 수 있도록 하여야 한다.

## (3) 시험 점수 중 최근 시험에 더 높은 비중을 주고 싶을 때
#      - 가중 평균
# 이유 : 최근 시험에 가중치를 크게 주어 평균을 내면
#        더 높은 비중을 주고 싶다는 의도에 맞게 사용할 수 있다.
```

```python
## ===============================================================
## [문제 2] 반별 평균을 통합할 때 – 가중평균
## ===============================================================
## A반과 B반의 시험 결과가 다음과 같다.
##
## A반: 학생 수 10명, 평균 90점
##
## B반: 학생 수 40명, 평균 70점

## 전체 50명에 대한 단순 평균을 잘못 계산한 경우 (반 평균의 평균)과
##
## 올바른 전체 평균(가중평균) 을 계산하세요.

# 단순 평균으로 잘못 계산한 경우
result1 = (90+70)/2
print(result1)  ## 더 높은 학생 수가 있는 반에 가중을 주지 않음

# 가중 평균으로 계산한 경우
result2 = ( (10*90) + (40*70) ) / (10 + 40)
print(result2)  ## 더 높은 반에 가중을 주었음
```

```python
## ===============================================================
## [문제 3] 절사 평균 계산
## ===============================================================
## 데이터는 아래와 같습니다. 
## x=[10,12,11,9,13,200]
x=[10,12,11,9,13,200]
## 일반 평균을 구하시오.
result1 = sum(x)/len(x)
print(result1)

## 가장 작은 값과 가장 큰 값(10, 200)을 하나씩 제거한 후 절사 평균을 구하시오.
result2 = sum(x[1:-1])/len(x[1:-1])
print(result2)

## 왜 절사 평균이 일반 평균보다 극단값에 덜 민감한지 설명하시오.
# 양극에 있는 극단값들을 지웠는데, x라는 데이터의 가장 오른쪽에는 200이라는 극단값이 있어서, 이 값이 사라지고 극단값에 덜 민감하게 바뀐다.
```

```python
## ===============================================================
## [문제 4] 가중 평균 계산
## ===============================================================
## 어떤 회사의 고객 만족도 조사는 다음과 같습니다. 
##  그룹	응답자 수	평균 점수
##  VIP	       20	    4.8
##  일반	  200	    4.2
##
## 전체 고객의 가중 평균 만족도 점수를 계산하시오.
## 단순히 (4.8 + 4.2) / 2 로 계산하는 것이 왜 잘못되었는지 설명하시오.

result = ( (20*4.8)+(200*4.2) )/(20+200)
print(result)

# VIP의 응답자 수는 20명이고, 일반은 200명의 응답자 수에서 평균을 낸 것이므로 둘의 가중치는 달라야
# 더 정확한 값이 나오기 때문에 단순한 평균 계산은 잘못된 계산이다.
```

```python
## ===============================================================
##  [문제 5] Pandas로 그룹별 가중평균 계산 
## ===============================================================
## 다음과 같은 DataFrame이 있습니다.

##  df = pd.DataFrame({
##      "group":  ["A", "A", "A", "B", "B"],
##      "score":  [80, 90, 70, 60, 100],
##      "weight": [1, 2, 1, 3, 1]
##  })

import pandas as pd
df = pd.DataFrame({
     "group":  ["A", "A", "A", "B", "B"],
     "score":  [80, 90, 70, 60, 100],
     "weight": [1, 2, 1, 3, 1]
 })

## group별 단순 평균(score의 평균)을 구하시오.
result1 = df.groupby("group")["score"].mean()
print(f'단순 평균 : {result1}')

## group별 가중평균(weight를 가중치로 사용)을 구하는 Pandas 코드를 작성하시오.
num = (df["score"] * df["weight"]).groupby(df["group"]).sum()
den = df["weight"].groupby(df["group"]).sum()
result2 = num / den

print(f'가중 평균 : {result2}')

```

```python
## ===============================================================
##  [ 문제 6 ] 분산과 표준편차 비교
## ===============================================================
## 다음은 어느 반 학생 10명의 수학 시험 점수이다.
## A반: [70, 72, 75, 75, 76, 80, 82, 85, 88, 90]
## B반: [40, 55, 60, 70, 75, 90, 95, 98, 100, 100]

class_A = [70, 72, 75, 75, 76, 80, 82, 85, 88, 90]
class_B = [40, 55, 60, 70, 75, 90, 95, 98, 100, 100]

df = pd.DataFrame({
    "class": ["A"] * 10 + ["B"] * 10,
    "score": class_A + class_B
})

## 두 반의 분산(sample variance)을 각각 구하시오.
result1 = df.groupby('class')['score'].var()
print(result1)

## 두 반의 표준편차를 비교하고, 어느 반의 점수 분포가 더 불안정한지 설명하시오.
result2 = df.groupby('class')['score'].std()
print(result2)

# B반의 편차가 A반보다 약 3배정도 크므로 B반의 점수 분포가 더 불안정하다는 것을 알 수 있다.
```

```python
## ===============================================================
##  [ 문제 7] 중위절대편차(Median Absolute Deviation, MAD) 계산
## ===============================================================
## 다음은 특정 제조 공정에서 측정한 부품 길이(mm) 데이터입니다..

## [99.8, 100.1, 100.0, 100.2, 99.9, 100.3, 100.5, 94.0]
import numpy as np
length = [99.8, 100.1, 100.0, 100.2, 99.9, 100.3, 100.5, 94.0]

df = pd.DataFrame({'길이':length})

## 위 데이터의 중앙값(median)을 구하시오.
median_value = df['길이'].median()
print(f'중앙값 : {median_value}')

## 각 데이터에서 중앙값을 뺀 절대값을 구하시오.
abs_value = abs(df['길이'] - median_value)
print(abs_value)


## 그 절대값 리스트의 중앙값(MAD)을 구하시오.
MAD1 = abs_value.median()
print(f'절대값 리스트의 중앙값 : {MAD1}')

## 이 데이터에서 MAD가 낮은지 높은지 해석하시오.
# => 데이터의 대부분의 중앙값 100.05 근처에 있으므로 MAD 0.2는 낮은값이다.

## 마지막 데이터 94.0을 제거했을 때 MAD는 어떻게 변하는가? 그 이유는?
length2 = [99.8, 100.1, 100.0, 100.2, 99.9, 100.3, 100.5]
df2 = pd.DataFrame({'길이': length2})

median2 = df2['길이'].median()
abs_dev2 = abs(df2['길이'] - median2)
MAD2 = abs_dev2.median()
print(MAD2)

# => MAD는 원래 이상치에 강한 지표라서 마지막 데이터인 94.0을 제거해도 크게 변하지 않는다.
```

```python
## ===============================================================
##  [ 문제 8 ] 이상치 포함 시 통계량 비교
## ===============================================================
## 어느 판매원의 하루 판매 금액(만원 단위)이 일주일 동안 다음과 같습니다.
##  [10, 12, 11, 9, 10, 13, 100]

day_price = [10, 12, 11, 9, 10, 13, 100]

df = pd.DataFrame({'판매 금액':day_price})


##  평균(mean)과 중앙값(median)을 각각 구하시오.
print(df['판매 금액'].mean())
print(df['판매 금액'].median())

##  분산과 표준편차를 구하시오.
print(df['판매 금액'].var())
print(df['판매 금액'].std())

##  중위절대편차(MAD)를 구하시오.
median_val = df['판매 금액'].median()
abs_dev = abs(df['판매 금액'] - median_val)
MAD = abs_dev.median()
print(MAD)


##  왜 이 데이터에서는 평균보다 중앙값이 대표값으로 적절한지 설명하시오.
# => 데이터 값에 100이라는 이상치가 존재하여서 평균 값이 이상하게 튀었기 때문에 중앙값이 더 적절하다

##  MAD가 왜 robust한 지표인지 해당 데이터를 기준으로 설명하시오.
# => MAD는 중앙값 기준으로 편차를 계산하기 때문에 100처럼 극단적으로 큰 값은 영향을 받지 않아서 이상치가 있어도 변하지 않는다.


```

```python
## ===============================================================
##  [ 문제 9] 상자그림(Boxplot) 분석
## ===============================================================
## 다음은 두 공정(A, B)에서 생산된 동일 부품의 길이 데이터이다.

## A공정: [50, 50, 51, 50, 49, 50, 51, 52]
## B공정: [50, 55, 45, 50, 60, 48, 54, 52]

A = [50, 50, 51, 50, 49, 50, 51, 52]
B = [50, 55, 45, 50, 60, 48, 54, 52]

df = pd.DataFrame({'A_length':A, 'B_length':B})

## 각 공정의 Q1, Q2(중앙값), Q3를 구하시오.
A_q1 = df['A_length'].quantile(.25)
A_q2 = df['A_length'].quantile(.50)
A_q3 = df['A_length'].quantile(.75)
print(f'A공정 : {A_q1}, {A_q2}, {A_q3}')

B_q1 = df['B_length'].quantile(.25)
B_q2 = df['B_length'].quantile(.50)
B_q3 = df['B_length'].quantile(.75)
print(f'B공정 : {B_q1}, {B_q2}, {B_q3}')

## IQR(사분위수 범위)을 계산하시오.
A_IQR = A_q3 - A_q1
B_IQR = B_q3 - B_q1

## 사분위수 1.5 IQR 규칙에 따라 각 공정의 이상치를 판정하시오.
A_lower = A_q1 - 1.5 * A_IQR
A_upper = A_q3 + 1.5 * A_IQR
cond_A = (df['A_length'] < A_lower) | (df['A_length'] > A_upper)
A_outliers = df['A_length'][cond_A]
print(A_outliers) # 빈 리스트 => 이상치 없음 

B_lower = B_q1 - 1.5 * B_IQR
B_upper = B_q3 + 1.5 * B_IQR
cond_B = (df['B_length'] < B_lower) | (df['B_length'] > B_upper)
B_outliers = df['B_length'][cond_B]
print(B_outliers) # 빈 리스트 => 이상치 없음 


## 두 공정의 상자그림을 그리고, 변동성 차이를 해석하시오.
import matplotlib.pyplot as plt
import koreanize_matplotlib
plt.boxplot([df['A_length'], df['B_length']], labels=['A공정', 'B공정'])
plt.title('A공정 vs B공정 Boxplot')
plt.ylabel('부품 길이')
plt.grid(True)
plt.show()

## 품질 안정성 측면에서 어떤 공정이 더 적합한지 설명하시오.
# A공정이 IQR이 작고 변동폭이 B에 비해 좁기 때문에 안정성 측면에서 더 적합한 공정이다!
```

```python
## ===============================================================
##  [ 문제 10 ] 분산·표준편차 vs MAD vs Boxplot 통합 분석
## ===============================================================
## 다음은 20명의 하루 걸음수 데이터입니다.

## [3500, 4200, 3800, 4000, 4100,
##  3950, 4020, 3900, 3800, 3700,
##  20000, 21000, 20500, 19800, 22000,
##  3600, 3700, 3800, 3900, 4000]

walk = [3500, 4200, 3800, 4000, 4100,
 3950, 4020, 3900, 3800, 3700,
 20000, 21000, 20500, 19800, 22000,
 3600, 3700, 3800, 3900, 4000]

df = pd.DataFrame({'걸음수':walk})

## 평균, 중앙값을 구하시오.
print('평균 :', df['걸음수'].mean())
print('중앙값 :', df['걸음수'].median())

## 분산, 표준편차를 구하시오.
print('분산 :', df['걸음수'].var())
print('표준편차 :', df['걸음수'].std())

## MAD를 구하시오.
median_val = df['걸음수'].median()
abs_dev = abs(df['걸음수'] - median_val)
MAD_val = abs_dev.median()
print('MAD :', MAD_val)

## 상자그림(Boxplot)을 그리고 극단치를 판정하시오.

# 극단치 먼저 판정해보고 비교해보기
q1 = df['걸음수'].quantile(0.25)
q3 = df['걸음수'].quantile(0.75)
IQR = q3 - q1
lower = q1 - 1.5*IQR  
upper = q3 + 1.5*IQR
cond = (df['걸음수'] < lower) | (df['걸음수'] > upper)
outliers = df['걸음수'][cond]
print(outliers)


# 상자그림 그리기
import matplotlib.pyplot as plt
plt.boxplot(df['걸음수'])
plt.title("걸음수 Boxplot")
plt.ylabel("걸음 수")
plt.grid(True)
plt.show()



## 이 데이터에서 “변동성을 판단할 때 표준편차보다 MAD가 더 적합한 이유”를 설명하시오.
# => 표준편차는 이상치가 들어오면 값이 이상해지는데 MAD는 중앙값 기반이라서 이상치 영향이 거의 없다.
#    따라서 극단값이 포함된 걸음수라는 데이터에서는 MAD가 더 적합하다.

## 만약 건강앱에서 사용자 걸음 트렌드를 분석하려 한다면,
## 어떤 대표값과 어떤 변동성 지표를 사용하는 것이 좋은지 이유와 함께 제안하시오.
# => 대표값     : 중앙값
#    변동성 지표 : MAD
# ===> 중앙값은 이상치의 영향을 받지 않기 때문에 대표값으로는 중앙값이 좋고,
#      극단값이 존재하는 걸음 수에서 값이 흔들리지 않고 규칙성을 파악하기에는 MAD가 좋다.
```
