### 【 기술통계/요약통계 - 데이터의 특성을 수치로 설명하는 것 】

[1] 모듈 로딩 및 데이터 준비 <hr>

```python
## 모듈 로딩
import pandas as pd
import numpy as np
import wquantiles as wq
```

```python
## 임의의 데이터
dataA = [70, 72, 75, 75, 76, 80, 82, 85, 88, 90]
dataSR = pd.Series(dataA)
```

```python
## 기술통계값 확인
dataSR.describe(percentiles=[0.1, 0.9])
```

```python
## 평균/분산/표준편차
## => pandas
df = pd.DataFrame({'점수':dataA})
print('평균 :',df.mean())
print('분산 :',df.var())
print('표준편차 :',df.std())


## => numpy
```
