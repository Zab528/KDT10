- 1-2: 컬럼명 리스트를 `q1_2_cols`, 인덱스 RangeIndex의 시작/끝을 튜플 `q1_2_index_info`로.

```python
# === [1] DataFrame 기초 생성 및 속성 이해 ===
# 아래 변수들에 결과를 담으세요.
# 1-1
q1_1 = pd.read_csv('../EXAM_DATA/people_basic.csv')
display(q1_1)
# 1-2
q1_2_cols = q1_1.columns.tolist
print(q1_2_cols)

q1_2_index_info = q1_1.index[:]
print(q1_2_index_info)

# 1-3
q1_3_rows = len(q1_1.index)
print(q1_3_rows)

q1_3_cols = len(q1_1.columns)
print(q1_3_cols) 

# 1-4
q1_4_stats = (q1_1['age'].mean() , q1_1['age'].median())
print(q1_4_stats)

# 1-5
q1_5_df = q1_1[q1_1['age']>=30]
print(q1_5_df)
```

<hr> 3. 조건 필터링과 정렬 (cars_mpg) <hr> 

- 3-1: mpg >= 30 인 행만 `q3_1_df`에.
- 3-2: manufacturer == 'audi' & year == 1999 인 행만 `q3_2_df`에.
- 3-3: horsepower 내림차순 정렬 결과를 `q3_3_sorted`에.
- 3-4: origin별 평균 mpg Series를 `q3_4_mean_mpg`에.
- 3-5: mpg 상위 5% 차량만 `q3_5_top5p`에.

```python
# === [3] 조건 필터링과 정렬 ===
DATA_FILE = '../EXAM_DATA/cars_mpg.csv'
df = pd.read_csv(DATA_FILE)

# 3-1
q3_1_df = df[df['mpg']>=30]
display(q3_1_df)

# 3-2
mask1 = df['manufacturer'] == 'audi'
mask2 = df['year'] == 1999
q3_2_df = df[mask1 & mask2]
display(q3_2_df)

# 3-3
q3_3_sorted = df.sort_values('horsepower',ascending=False)
display(q3_3_sorted)

# 3-4
q3_4_mean_mpg = df.groupby('origin')['mpg'].mean()
display(q3_4_mean_mpg)

# 3-5
# 19개에서 상위 5% => 가장 큰 1개
q3_5_top5p = df.sort_values('mpg', ascending=False).iloc[0]
display(q3_5_top5p)
```

<hr> 4. 결측치 처리 (titanic_simple) <hr> 

- 4-1: 컬럼별 결측치 개수 Series를 `q4_1_na`에.
- 4-2: age 결측치를 평균으로 채운 DataFrame을 `q4_2_filled`에.
- 4-3: 결측치가 1개라도 있는 행 제거 DataFrame을 `q4_3_dropna`에.
- 4-4: 결측치 여부(True/False) DataFrame을 `q4_4_isna`에.
- 4-5: 처리 전후 age 평균을 튜플로 `q4_5_before_after`에.

```python
# === [4] 결측치 처리 ===
DATA_FILE = '../EXAM_DATA/titanic_simple.csv'
df = pd.read_csv(DATA_FILE)
display(df)

# 4-1
q4_1_na = ({'name', 'age', 'sex', 'survived'},
           {df['name'].isna().sum(), df['age'].isna().sum(), df['sex'].isna().sum(), df['survived'].isna().sum()})
q4_1_na = pd.DataFrame(q4_1_na)
display(q4_1_na)

# 4-2
q4_2_filled = df.fillna(df['age'].mean())
display(q4_2_filled)

# 4-3
q4_3_dropna = df.dropna()
display(q4_3_dropna)

# 4-4
q4_4_isna = df.isna()
display(q4_4_isna)

# 4-5
q4_5_before_after = (df['age'].mean() , q4_2_filled['age'].mean())
print(q4_5_before_after)
```

<hr> 5. 그룹화와 집계 (employees) <hr> 

- 5-1: city 대신 employees 사용: department별 평균 salary Series를 `q5_1_dept_mean`에.
- 5-2: gender별 최대 salary Series를 `q5_2_gender_max`에.
- 5-3: (department, rank) 기준으로 salary 평균을 DataFrame `q5_3_group_mean`에.
- 5-4: salary에 대해 평균/중앙값/표준편차를 agg로 계산한 DataFrame을 `q5_4_agg`에.
- 5-5: department별 데이터 개수를 오름차순으로 정렬한 Series를 `q5_5_counts_sorted`에.

```python
# === [5] 그룹화와 집계 ===
DATA_FILE = '../EXAM_DATA/employees.csv'
df = pd.read_csv(DATA_FILE)
display(df)

# 5-1
q5_1_dept_mean = df.groupby('department')['salary'].mean()
display(q5_1_dept_mean)

# 5-2
q5_2_gender_max = df.groupby('gender')['salary'].max()
display(q5_2_gender_max)

# 5-3
q5_3_group_mean = df.groupby(['department','rank'])['salary'].mean()
display(q5_3_group_mean)

# 5-4
q5_4_agg = df['salary'].describe()
display(q5_4_agg)

# 5-5
q5_5_counts_sorted = df.groupby('department')
print(q5_5_counts_sorted)
```
