```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/inf/refs/heads/main/p1/delivery_time.csv")

# Section 25. 시간 간의 차이 계산(일)
display(df)

# 각 사용자별로 첫 주문과 마지막 주문 사이의 시간 간격을 일 단위로 계산하시오. ## 깍두기
df['주문시간'] = pd.to_datetime(df['주문시간'])
min_order_time = df.groupby('user')['주문시간'].min()
max_order_time = df.groupby('user')['주문시간'].max()
answer = ((max_order_time) - (min_order_time)).dt.days
answer

# 한 번만 주문했거나, 같은 날 주문이 이루어진 사용자(시간 간격이 0일인 사용자)는 제외하고, 나머지 사용자들의 평균 시간 간격(일 단위)을 계산하시오. ## 깍두기
cond = answer>0
answer2 = answer[cond].mean()
answer2

# 평균 시간 간격보다 긴 시간 간격을 가진 사용자의 수를 정수로 구하시오. ## 도전
cond = answer > answer2
print(int(sum(cond)))
```

### Section29.

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/inf/refs/heads/main/p1/delivery_time.csv")

# Section 29. 함수, 월별 집계
def delivery_fee(distance):
    if distance < 5:
        return 2000
    elif distance < 10:
        return 4000
    elif distance < 15:
        return 6000
    elif distance < 20:
        return 8000

# 배달료 계산 기준표에 따라 각 주문에 대한 배달료를 계산하시오. #깍두기
df['배달요금'] = df['거리'].apply(delivery_fee)

# 월별로 배달료의 총합을 집계하시오.
df['주문시간'] = pd.to_datetime(df['주문시간'])
df['month'] = df['주문시간'].dt.month
df_new = df.groupby(['month'])['배달요금'].sum()
df_new

# 배달료가 가장 많이 발생한 월을 찾고, 그 월의 총 배달료를 정수로 구하시오.
df_new = df_new.sort_values(ascending=False)
print(int(df_new.iloc[0]))

```

### Section37.

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert_v2/refs/heads/main/part1/ch3/store_sales.csv")

# Section 37. 재구조화 (unstack), 맵핑
df

# 각 행에서 판매수량과 단가를 이용하여 매출액을 계산하세요.
df['매출액'] = df['판매수량'] * df['단가']
df

# 요일을 평일과 주말로 구분하고, 매장별 평일과 주말 매출액 합계를 구하시오.     # 깍두기 : 요일 매핑하는 방법
day_mapping = {'월':'평일', '화':'평일', '수':'평일', '목':'평일', '금':'평일',
               '토':'주말', '일':'주말'}
df['요일구분'] = df['요일'].map(day_mapping)
df

# 매장별 평일과 주말 매출액 차이를 절대값으로 구하세요. 이후, 모든 매장 중 가장 큰 절대값 차이를 찾으시오.
df_new = df.groupby(['매장코드','요일구분'])['매출액'].sum().unstack()
df_new['차이'] = abs(df_new['주말']-df_new['평일'])
df_new = df_new.sort_values('차이',ascending=False)
print(df_new['차이'].iloc[0])
```

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert_v2/refs/heads/main/part1/ch3/region_sales.csv")

# Section 38. 피벗테이블
df

# 각 Region과 Channel 조합별로 제품(Product) A, B의 총 판매액을 계산하시오.     # 깍두기 : 피벗 테이블 만들기
pt = pd.pivot_table(df, index=['Region', 'Channel'], columns='Product',
                            values='Sales', aggfunc='sum')
pt

# 제품 A의 매출 비율(A비율)을 구하세요. A비율 = (제품 A 판매액) ÷ (제품 A 판매액 + 제품 B 판매액)   # 깍두기 : 매출 비율 계산하기
pt['총 매출'] = pt['A'] + pt['B']
pt['A 비율'] = pt['A'] / pt['총 매출']
pt

# A비율 중 최댓값을 찾아 반올림하여 소수 둘째 자리까지 구하시오.
round(pt['A 비율'].max(),2)


```
