```python
import pandas as pd
```

```python
## ------------------------------------------------------------
# 문제 1. 아래 데이터를 저장 후 조건을 만족하는 코드를 작성하세요.
## ------------------------------------------------------------
data = [
     ["2차전지(생산)", "SK이노베이션", 10.19, 1.29],
     ["해운", "팬오션", 21.23, 0.95],
     ["시스템반도체", "티엘아이", 35.97, 1.12],
     ["해운", "HMM", 21.52, 3.20],
     ["시스템반도체", "아이에이", 37.32, 3.55],
     ["2차전지(생산)", "LG화학", 83.06, 3.75]
 ]
columns = ["테마", "종목명", "PER", "PBR"]

dataDF = pd.DataFrame(data, columns=columns)
display(dataDF)
```

```python
#  (1-1) 종목명, PER, PBR 컬럼만 선택 후 DataFrame으로 저장하는 코드 작성하세요.
dataDF1 = dataDF.loc[:, ['종목명','PER','PBR']]
display(dataDF1)
```

```python
#  (1-2)  PER이 20미만, 50이상인 데이터만 저장하는 코드 작성하세요.
dataDF2 = dataDF[(dataDF['PER'] < 20) | (dataDF['PER'] >= 50)]
display(dataDF2)
```

```python
#  (1-3) 종목명을 행인덱스로 설정하는 코드 작성하세요.
dataDF3 = dataDF.set_index(['종목명'])

display(dataDF3)
```

```python
#  (1-4) PBR 컬럼의 데이터를 기준으로 내림차순 정렬해 주세요.
dataDF4 = dataDF.sort_values('PBR')
display(dataDF4)
```

```python
#  (1-5) PER, PBR 순서로 오름차순 정렬해 주세요.
dataDF.sort_values(by = ['PER','PBR'], axis='index', ascending=True)
```

```python
## ------------------------------------------------------------
# 문제 5. 아래 데이터를 저장 후 조건을 만족하는 코드를 작성하세요.
## ------------------------------------------------------------
# 데이터파일 :  titanic.csv
FILE_NAME = '../Data/titanic.csv'
dataDF = pd.read_csv(FILE_NAME)
dataDF
```

```python
# (5-1) 단순 조건에 맞는 데이터만 선택 
# ▶ 성별이 여자인 승객 행만 필터링.
# ▶ pclass가 1등석인 승객만.
# ▶ age가 10세 미만인 어린이만.
# ▶ fare가 100이상인 고가 요금 승객만.
dataDF[(dataDF['sex'] == 'female') & (dataDF['pclass'] == 1) & (dataDF['age'] < 10) & (dataDF['fare'] >= 100)]
```

```python
# (5-2) 아래 조합 조건에 맞는 데이터만 선택
# ▶ 1 또는 2등석인 생존자 데이터만
dataDF[(dataDF['survived'] == 1) & ((dataDF['pclass'] == 1) | (dataDF['pclass'] == 2))]
```

```python
# ▶ 가족 동승 3명 이상) 승객 데이터만
dataDF[(dataDF['sibsp'] >= 3)]
```

```python
# ▶ 혼자 탑승한 성인 남자 데이터만
dataDF[(dataDF['adult_male'] == True) & (dataDF['sibsp'] == 0)]
```
