### 【 중복 데이터 체크 및 처리 】
- 중복 데이터 여부 검사 : DF.duplicated()  => True/False
- 중복 데이터 삭제 처리 : DF.drop_duplicates()

```python
## 모듈로딩
import pandas as pd

## 데이터 준비

df= pd.DataFrame({
    'brand' : [ 'Yum', 'Yum', 'Indo', 'Indo', 'Indo' ],
    'style'   : [ 'cup', 'cup', 'cup', 'pack', 'pack' ],
    'rating' : [4, 4, 3.5, 15, 5]
})

display(df)
```

```python
## 중복체크
result = df.duplicated().sum()

print(f'중복 데이터/행 : {result}개')
```

```python
## 중복체크
## - keep 매개변수 : 'first' 첫 번째 유지, 나머지 중복
result= df.duplicated()
result
```

```python
## 중복체크
## - keep 매개변수 : 'first' 첫 번째 유지, 나머지 중복
result= df.duplicated(keep='last')
result
```

```python
## 중복체크 - subset매개변수 : 중복 검사 컬럼 설정

## brand 컬럼만 중복인 행
result = df.duplicated(subset=['brand', 'style'])   ## keep='first'
result
```

```python
## 중복데이터 삭제 - 모든 속성값이 동일한 행 비교
df.drop_duplicates()
```

```python
## 중복데이터 삭제 - 모든 속성값이 동일한 행 비교
df.drop_duplicates(keep='last')
```

```python
## brand 컬럼만 중복인 행
df.drop_duplicates(subset=['brand', 'style'], keep='last') 
```
