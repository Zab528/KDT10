#### 【 DataFrame 연산 】
- Series와 동일한 원리로 연산
- 같은 위치에 있는 원소끼리 연산
- 종류
    * DataFrame객체와 숫자 연산
    * DataFrame객체와 DataFrame객체 연산
    * DataFrame객체와 연산메서드

[1] 모듈 로딩 및 데이터 준비 <hr>

```python
## 모듈 로딩
import pandas as pd
```

```python
## 데이터프레임 생성
dataDF = pd.Series([10, 20, 30],
                   [40, 50, 60])
dataDF
```

[2] DataFrame와 숫자 연산 <hr>

```python
## 원소 단위로 각각 계산 진행 ==> 결과: DataFrame
ret1 = dataDF + 17
ret2 = dataDF - 17
ret3 = dataDF * 17
ret4 = dataDF / 17
ret5 = dataDF % 17

print(f'dataDF + 17 =>\n{ret1}\n')
print(f'dataDF + 17 =>\n{ret2}\n')
print(f'dataDF + 17 =>\n{ret3}\n')
print(f'dataDF + 17 =>\n{ret4}\n')
print(f'dataDF + 17 =>\n{ret5}\n')
```
