### [ 파이썬에서 변수와 참조 ]
- 데이터 저장 문법 : 변수명 = 데이터
- 데이터 저장 의미
    * 힙메모리 영역에 저장된 데이터(객체)의 주소를 저장
    * 스택메모리 영역에 변수명 존재함 => 참조형 변수(reference variable)


```python
a = 10       ## int형 인스턴스 10의 주소를 변수a에 저장
b = 10       ## int형 인스턴스 10의 주소를 변수b에 저장
c =  a       ## int형 인스턴스 10의 주소를 변수c에 저장

print('id(a)', id(a))
print('id(b)', id(b))
print('id(c)', id(c))
```

```python
a = 5
b = a
print('id(a)', id(a))
print('id(b)', id(b))

a += 5                  ## a = a+5 => 5 + 5 = 10  : 10의 주소
print('id(a)', id(a))
print('id(b)', id(b))
```

```python
import pandas as pd

sr00 = pd.Series([11,22], index=['a','b'])   ## 힙에 시리즈 객체 주소 저장

sr11 = sr00                                  ## 힙에 시리즈 객체 주소 저장

print('----------변경전----------')
print('id(sr11)', id(sr11), sr11.index)
print('id(sr00)', id(sr00), sr00.index)

sr00.index=[11,22]
print('----------변경후----------')
print('id(sr11)', id(sr11), sr11.index)
print('id(sr00)', id(sr00), sr00.index)

## = 복사하기 copy()-------------
sr22 = sr00.copy()
print('----------변경전----------')
print('id(sr22)', id(sr22), sr22.index)
print('id(sr00)', id(sr00), sr00.index)

sr00.index=['하나','둘']
print('----------변경후----------')
print('id(sr22)', id(sr22), sr22.index)
print('id(sr00)', id(sr00), sr00.index)
```
