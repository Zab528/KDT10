### 【 데이터 시각화 】
- 데이터셋 : iris.csv
- 해결문제 : 품종별 시각화에 적합한 특성/속성/컬럼 선정
- 출력결과 : 선정된 특성을 기반으로 품종 분류한 것 시각화

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../Data/iris.csv')
df.isnull().sum()
```

```python
df
```

```python
plt.plot(df['variety'])
```

```python
df_se = df[df['variety'] == 'Setosa']
df_ve = df[df['variety'] == 'Versicolor']
df_vi = df[df['variety'] == 'Virginica']

df_se.index = range(1, len(df_se)+1)
df_ve.index = range(1, len(df_ve)+1)
df_vi.index = range(1, len(df_vi)+1)
```

```python
fig, axes = plt.subplot_mosaic([['top_left', 'top_right'],
                                ['bottom_left', 'bottom_right']])


## Setosa의 4개의 컬럼 값들 출력

axes['top_left'].plot(df_se.index, df_se['sepal.length'])
axes['top_left'].set_title('Sepal Length')

axes['top_right'].plot(df_se.index, df_se['sepal.width'])
axes['top_right'].set_title('Sepal Width')

axes['bottom_left'].plot(df_se.index, df_se['petal.length'])
axes['bottom_left'].set_title('Petal Length')

axes['bottom_right'].plot(df_se.index, df_se['petal.width'])
axes['bottom_right'].set_title('Petal Width')
```

```python
fig, axes = plt.subplot_mosaic([['top_left', 'top_right'],
                                ['bottom_left', 'bottom_right']])

## Vesicolor의 4개의 컬럼 값들 출력

axes['top_left'].plot(df_ve.index, df_ve['sepal.length'])
axes['top_left'].set_title('Sepal Length')

axes['top_right'].plot(df_ve.index, df_ve['sepal.width'])
axes['top_right'].set_title('Sepal Width')

axes['bottom_left'].plot(df_ve.index, df_ve['petal.length'])
axes['bottom_left'].set_title('Petal Length')

axes['bottom_right'].plot(df_ve.index, df_ve['petal.width'])
axes['bottom_right'].set_title('Petal Width')
```

```python
fig, axes = plt.subplot_mosaic([['top_left', 'top_right'],
                                ['bottom_left', 'bottom_right']])

## Virginica의 4개의 컬럼 값들 출력

axes['top_left'].plot(df_vi.index, df_vi['sepal.length'])
axes['top_left'].set_title('Sepal Length')

axes['top_right'].plot(df_vi.index, df_vi['sepal.width'])
axes['top_right'].set_title('Sepal Width')

axes['bottom_left'].plot(df_vi.index, df_vi['petal.length'])
axes['bottom_left'].set_title('Petal Length')

axes['bottom_right'].plot(df_vi.index, df_vi['petal.width'])
axes['bottom_right'].set_title('Petal Width')
```

```python
fig, axes = plt.subplot_mosaic([['top_left', 'top_right'],
                                ['bottom_left', 'bottom_right']])

# 1. sepal.length
axes['top_left'].plot(df_se.index, df_se['sepal.length'], label='Setosa')
axes['top_left'].plot(df_ve.index, df_ve['sepal.length'], label='Versicolor')
axes['top_left'].plot(df_vi.index, df_vi['sepal.length'], label='Virginica')
axes['top_left'].set_title('Sepal Length')
axes['top_left'].legend(loc='best')

# 2. sepal.width
axes['top_right'].plot(df_se.index, df_se['sepal.width'], label='Setosa')
axes['top_right'].plot(df_ve.index, df_ve['sepal.width'], label='Versicolor')
axes['top_right'].plot(df_vi.index, df_vi['sepal.width'], label='Virginica')
axes['top_right'].set_title('Sepal Width')
axes['top_right'].legend(loc='best')

# 3. petal.length
axes['bottom_left'].plot(df_se.index, df_se['petal.length'], label='Setosa')
axes['bottom_left'].plot(df_ve.index, df_ve['petal.length'], label='Versicolor')
axes['bottom_left'].plot(df_vi.index, df_vi['petal.length'], label='Virginica')
axes['bottom_left'].set_title('Petal Length')
axes['bottom_left'].legend(loc='best')

# 4. petal.width
axes['bottom_right'].plot(df_se.index, df_se['petal.width'], label='Setosa')
axes['bottom_right'].plot(df_ve.index, df_ve['petal.width'], label='Versicolor')
axes['bottom_right'].plot(df_vi.index, df_vi['petal.width'], label='Virginica')
axes['bottom_right'].set_title('Petal Width')
axes['bottom_right'].legend(loc='best')

plt.tight_layout()
plt.show()
```
