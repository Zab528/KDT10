### 【 데이터 시각화 - 공통사항 】
- 데이터 비교 시 많이 사용되는 그래프
- 정렬 시켜서 시각화시 빠르게 인식 가능
- 데이터따라서 수평/수직으로 설정해서 시각화
- 함수 : matplotlib.pyplot.bar() / barh()

```python
### [1] 모듈 로딩 
import matplotlib.pyplot as plt
import koreanize_matplotlib
import pandas as pd
```

```python
## ===================================================
## [2] 기본 막대 그래프
## ===================================================
## - 데이터
xData = [10, 20, 30, 40, 50]
yData = [10, 20, 30, 40, 50]

## - 막대 그래프 인스턴스 생성 : 기본값으로 선 그래프 생성
line = plt.bar(range(1, len(xData)+1), yData)

## - 그래프 기본 요소 설정
plt.title('기본 막대그래프')
plt.xlabel('X 데이터')
plt.ylabel('Y 데이터')

## - 화면에 그래프 출력
plt.show()
```

```python
## ===================================================
## [2] 기본 막대 그래프 + DataFrame
## ===================================================

## - 데이터
dataDF = pd.DataFrame( {'A': [10, 50, 30, 40, 20],
                        'B': [15, 35, 45, 25, 55]})
display(dataDF)

dataDF = dataDF.sort_values(by=['A'], ascending=False).reset_index(drop=True)
display(dataDF)

## - 막대 그래프 인스턴스 생성 : 기본값으로 선 그래프 생성
bar = plt.bar(dataDF.A.index, dataDF.A)

## - 그래프 기본 요소 설정
plt.title('기본 막대 그래프')
plt.xlabel('X 데이터')
plt.ylabel('Y 데이터')

## - 화면에 그래프 출력
plt.show()
```

```python
## ===========================================================================
## [2] 사용자 설정 그래프 - 타이틀/ x라벨 / y라벨
## ===========================================================================
## plt.title()  : 그래프 제목 설정
## plt.xlabel() : x축 라벨 설정
## plt.ylabel() : y축 라벨 설정
## 공통 속성     : 텍스트 출력에 따른 텍스트 속성 (글자크기, 색상, 두께...) 설정
## ===========================================================================

## - 선 그래프 인스턴스 생성 : 기본값으로 선 그래프 생성
##         plot(x, y)        : x축 데이터, y축 데이터
line = plt.plot(['a', 'b', 'c', 'd', 'e'], [10, 20, 30, 40, 50])

## 그래프 공통 요소 설정 => 각 요소별 인스턴스 반환 
fontdict  = {'size': 20, 'color':'red', 'weight':'bold'}
labeldict = {'size': 17, 'color':'blue', 'labelpad':15}

## -> 그래프와 타이틀 간격 pad 매개변수 설정 
## -> 그래프에 대한 타이틀 위치 loc 매개견수 설정 : 'left', 'center', 'right'
plt.title("[선 그래프]", loc='left', pad=20, **fontdict)

## -> 눈금과 라벨들의 간격 조절 labelpad 매개변수 설정 
## -> 라벨들의 위치 loc 매개견수 설정 : x, y라벨에 따라 값 다름
plt.xlabel("종류", loc='left', **labeldict)      ## 'left', 'center', 'right'
plt.ylabel("값",   loc='top',  **labeldict)      ## 'top', 'center', 'bottom'

## - 화면에 그래프 출력
plt.show()  
```

```python
## ===========================================================================
## [3] 사용자 설정 선 그래프  - 범례 설정 
## ===========================================================================
## 여러개 그래프 식별을 위한 범례 설정 
## -> 방법1: 그래프에 label속성 설정 + plt.legend()
## ===========================================================================

## - 그래프 인스턴스 생성
##   label 매개변수 설정 => 범례로 사용
line1 = plt.plot(['a', 'b', 'c', 'd', 'e'], [10, 20, 30, 40, 50], '+r', label='Line1')
line2 = plt.plot(['a', 'b', 'c', 'd', 'e'], [15, 25, 35, 45, 55], 'og', label='Line2')

## - 범례 화면 출력
plt.legend() 

## - 화면에 그래프 출력
plt.show()  
```

```python
## ===========================================================================
## [3] 사용자 설정 선 그래프  - 범례 설정 
## ===========================================================================
## 여러개 그래프 식별을 위한 범례 설정 
## -> 방법2: plt.legend(labels=[])
## ===========================================================================

## - 그래프 인스턴스 생성 
line1 = plt.plot(['a', 'b', 'c', 'd', 'e'], [10, 20, 30, 40, 50], '+r')
line2 = plt.plot(['a', 'b', 'c', 'd', 'e'], [15, 25, 35, 45, 55], 'og')

## 범례 관련 매개변수들 ---------------------------------------------------------
## 범례 화면 출력. labels 매개변수 : 그래스 수 만큼 리스트 담아 설정 
## 범례 배치 방식 매개변수 ncols=1 설정 : ncols=2
## 범례 출력 위치 매개변수 loc='best' 설정
## 범례 라벨 색상 매개변수 labelcolor='black' 설정 
## ---------------------------------------------------------------------------
fontdict  = {'fontsize':15, 'labelcolor':'blue' , 'shadow':True, 'title':'다양한 선그래프'}
plt.legend(labels=['L1','L2'], ncols=2, loc='best', **fontdict) 

## - 화면에 그래프 출력
plt.show()  
```

```python
## ===========================================================================
## [3] 사용자 설정 선 그래프  - 범례 설정 
## ===========================================================================
## 여러개 그래프 식별을 위한 범례 설정 
## -> 방법2: plt.legend(labels=[])
## ===========================================================================
## - 그래프 인스턴스 생성 
line1 = plt.plot(['a', 'b', 'c', 'd', 'e'], [10, 20, 30, 40, 50], '+r')
line2 = plt.plot(['a', 'b', 'c', 'd', 'e'], [15, 25, 35, 45, 55], 'og')

## 범례 관련 매개변수들 ---------------------------------------------------------
## 범례 박스 표시 여수 frameon 매개변수 : True/False
## 범례 배경색 설정 facecolor 매개변수  : 'red', 'lightgray', ...
## 범례 테두리 색상  edgecolor 매개변수 : 'red', 'lightgray', ...
## 라벨들 사이 간격 labelspacing 매개변수: float [기]0.5
## 박스 테두리와 내용 사이 안쪽 여백 borderpad 매개변수 :  float [기] 0.5
## ---------------------------------------------------------------------------
fontdict  = { 'fontsize':15, 'labelcolor':'blue' , 'shadow':True, 
              'title':'다양한 선그래프', 'frameon': True,
              'facecolor':'skyblue', 'edgecolor':'red', 
              'borderpad':0.5, 'labelspacing':1}

## - 범례 화면 출력
plt.legend(labels=['L1','L2'], ncols=2, loc='best', **fontdict) 

## - 화면에 그래프 출력
plt.show()  
```

```python
## ===========================================================================
## [4] 사용자 설정 선 그래프 - X, Y축 범위 설정 
## ===========================================================================
## X축, Y축 범위설정 xlim(left, right)/ ylim(bottom, top)/axis()
## plt.xlim((0, 5))             : X축의 범위: [xmin, xmax] 또는 (xmin, xmax)
## plt.ylim([0, 20])            : Y축의 범위: [ymin, ymax] 또는 (ymin, ymax)
## plt.axis([0, 5, 0, 20])      : X축, Y축 범위: [xmin, xmax, ymin, ymax] 또는 (xmin, xmax, ymin, ymax)
## plt.axis()                   : 현재 axes의 X축, Y축 범위값 반환
## ===========================================================================
line1 = plt.plot( [10, 20, 30, 40, 50], '+r', markersize=20)
line2 = plt.plot( [15, 25, 35, 45, 55], 'og', markeredgecolor='yellow')

## x, y축 범위 설정 
# plt.xlim((-1, 7))
# plt.ylim([0, 60])    
#plt.axis([-1, 6, 0, 60])
plt.axis('square')

## - 화면에 그래프 출력
plt.show()  
```

```python
## ===========================================================================
## [4] 사용자 설정 선 그래프 - X, Y축 숨기기 옵션 
## ===========================================================================
## X/Y 축,틱,라벨 표시/숨김 설정 : axis('on') 
## plt.axis('on')           : 'off' 그래프만 표시  
## ===========================================================================
line1 = plt.plot( [10, 20, 30, 40, 50], '+r', markersize=20)
line2 = plt.plot( [15, 25, 35, 45, 55], 'og', markeredgecolor='yellow')

plt.title("TEST")
plt.xlabel("AA")
plt.ylabel("BB")
plt.legend(labels=['A','B'])

## x, y축 범위 설정 
# plt.xlim((-1, 7))
# plt.ylim([0, 60])    
#plt.axis([-1, 6, 0, 60])
plt.axis('off')                     # 축, 눈금, 테두리 모두 숨김(그림만 보기 좋게)

## - 화면에 그래프 출력
plt.show()  

```

```python
## ===========================================================================
## [5] 사용자 설정 선 그래프 
## ===========================================================================
## X축, Y축 눈금관련 설정들 => 틱(Tick) 그래프의 축에 간격을 구분하기 위해 표시하는 눈금
## -> 리스트 또는 NumPy array 입력하면 눈금과 숫자 레이이 표시
## xticks(ticks, labels) : X축 눈금값과 라벨 설정 
## yticks(ticks, labels) : Y축 눈금값과 라벨 설정 
## tick_params() : 눈금 스타일 설정
## ===========================================================================

## - 그래프 출력 속성들 설정값 
propertys = {'ms':20, 'mec':'yellow', 'ls':'-.'}

## - 그래프 인스턴스 생성
line1 = plt.plot([1,2,3,4,5], [10, 20, 30, 40, 50], '+r', markersize=20)
line2 = plt.plot([1,2,3,4,5], [15, 25, 35, 45, 55], 'og', **propertys)

## - 눈금 설정
plt.xticks([1,3,6,9])
plt.yticks([5,25,50])

## - 화면에 그래프 출력
plt.show()  
```

```python
## ===========================================================================
## [5] 사용자 설정 선 그래프 - 틱(Tick) 눈금 설정
## ===========================================================================
## X축, Y축 눈금관련 설정들 => 틱(Tick) 그래프의 축에 간격을 구분하기 위해 표시하는 눈금
## -> 리스트 또는 NumPy array 입력하면 눈금과 숫자 레이이 표시
## xticks(ticks, labels) : X축 눈금값과 라벨 설정 
## yticks(ticks, labels) : Y축 눈금값과 라벨 설정 
## tick_params() : 눈금 스타일 설정
## ===========================================================================

## - 그래프 출력 속성들 설정값 
propertys = {'ms':20, 'mec':'yellow', 'ls':'-.'}

## - 그래프 인스턴스 생성
line1 = plt.plot([1,2,3,4,5], [10, 20, 30, 40, 50], '+r', markersize=20)
line2 = plt.plot([1,2,3,4,5], [15, 25, 35, 45, 55], 'og', **propertys)

## - 눈금 설정
plt.xticks([1,3,6,9], labels=['l1','l2', 'l3','l4'])
plt.yticks([5,25,50], labels=['L', 'M', 'T'])

## -> 눈금 제거
plt.xticks([], [])

## - 화면에 그래프 출력
plt.show()  
```

```python
## ===========================================================================
## [5] 사용자 설정 선 그래프 - 틱(Tick) 눈금 설정
## ===========================================================================
## X축, Y축 눈금관련 설정들 => 틱(Tick) 그래프의 축에 간격을 구분하기 위해 표시하는 눈금
## -> 리스트 또는 NumPy array 입력하면 눈금과 숫자 레이이 표시
## xticks(ticks, labels) : X축 눈금값과 라벨 설정 
## yticks(ticks, labels) : Y축 눈금값과 라벨 설정 
## tick_params() : 눈금 스타일 설정
## ===========================================================================

## - 그래프 출력 속성들 설정값 
propertys = {'ms':20, 'mec':'yellow', 'ls':'-.'}

## - 그래프 인스턴스 생성
line1 = plt.plot([1,2,3,4,5], [10, 20, 30, 40, 50], '+r', markersize=20)
line2 = plt.plot([1,2,3,4,5], [15, 25, 35, 45, 55], 'og', **propertys)

## 눈금 설정
plt.xticks([1,3,6,9], labels=['l1','l2', 'l3','l4'])
plt.yticks([5,25,50], labels=['L', 'M', 'T'])

## 눈금 스타일 설정 ----------------------------------------------
## axis         : 스타일 지정 축 설정  ‘x’, ‘y’, ‘both’
## direction    : 눈금 방향. ‘in’, ‘out’, ‘inout’
## length       : 눈금의 길이를 지정
## pad          : 눈금과 레이블과의 거리
## labelsize    : 레이블의 크기
## labelcolor   : 레이블의 색상
## top/bottom/left/right : 눈금 표시될 위치 True/False
## width        : 눈금의 너비
## color        : 눈금의 색상 
## ------------------------------------------------------------
plt.tick_params(axis='both', direction='inout', length=10, width=5, color='red',
                pad=15,   labelsize=14, labelcolor='green',
                bottom=True, right=True, left=False)

## - 화면에 그래프 출력
plt.show()  
```

```python
## ===========================================================================
## [6]] Figure & axes 속성 읽기
## ===========================================================================
## plt.gcf()    gcf = get current figure : 현재 figure 인스턴스 가져오기
## plt.gca()    gca = get current axes   : 현재 axes 인스턴스 가져오기
## ===========================================================================

## -그래프 인스턴스 생성 => 그래프가 들어갈 figure, axes를 생성
line1 = plt.plot([1,2,3,4,5], [10, 20, 30, 40, 50], '+r', markersize=20)

## - 현재 Figure/Axes 객체 가져오기
fig = plt.gcf()   # gcf = get current figure
ax  = plt.gca()   # gca = get current axes

## - 현재 Figure/Axes 객체 속성 설정 : set_속성명(값)
fig.set_facecolor('yellow')
fig.set_edgecolor('red')

## - Figure 관련 주요 속성 출력 : get_속성명()
print("=== [Figure 속성 정보] ===")
print(f"Figure 객체             : {fig}")
print(f"Figure 번호             : {fig.number}")
print(f"Figure 크기(inch)       : {fig.get_size_inches()}")
print(f"Figure DPI              : {fig.dpi}")
print(f"Figure 실제 픽셀 크기(px) : {fig.get_size_inches() * fig.dpi}")
print(f"Figure 배경색(facecolor) : {fig.get_facecolor()}")
print(f"Figure 가장자리색(edgecolor): {fig.get_edgecolor()}")
print(f"Figure tight_layout 여부: {fig.get_tight_layout()}")

## - Axes(좌표축) 속성도 함께 확인
print("\n=== [Axes 속성 정보] ===")
print(f"X축 범위: {ax.get_xlim()}")
print(f"Y축 범위: {ax.get_ylim()}")
print(f"X축 라벨: {ax.get_xlabel()}")
print(f"Y축 라벨: {ax.get_ylabel()}")
print(f"타이틀  : {ax.get_title()}")

plt.show()
```

```python
## ===========================================================================
## [6] Figure & axes 속성 읽기 
## ===========================================================================
## plt.figure(figsize=(10, 4), facecolor='lightgray', linewidth=0.0)
##            크기설정(W, H)
## ===========================================================================

## - Figure 생성 시 크기 설정 (단위: inch) : 가로 8인치, 세로 4인치
## - 모니터나 저장 시 실제 크기는 figsize * dpi로 계산 
## - 예: (8,4) × 100dpi → 800×400px
plt.figure(figsize=(10, 4), facecolor='lightgray', linewidth=0.5, frameon=True)   

## -  그래프 인스턴스 생성 
line1 = plt.plot( [1, 2, 3, 4, 5], [10, 20, 30, 40, 50], '+r',  markersize=20 )

## - Figure 속성 확인
fig = plt.gcf()
print("Figure size (inch):", fig.get_size_inches())
print("Figure dpi:", fig.dpi)
print("Pixel 크기(px):", fig.get_size_inches() * fig.dpi)

## - 출력
plt.show()
```
