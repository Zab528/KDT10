### 【 데이터 시각화 - Area Graph 】
- 선 그래프에서 추가된 기능
- 특정 영역을 강조/집중해서 표현하고 싶은 경우 사용
- 너무 많은 조각은 오히려 가독성 떨어지고 집중 X
- 함수 : matplotlib.pyplot.fill_between()/fill_betweenx()

```python
### [1] 모듈 로딩 
import matplotlib.pyplot as plt
import koreanize_matplotlib
import pandas as pd
```

```python
## ===================================================
## [2] 선 그래프
## ===================================================
## - 데이터
xData = [1, 2, 3, 4]
yData = [2, 3, 5,10]

## - 막대 그래프 인스턴스 생성 : 기본값으로 선 그래프 생성
line = plt.bar(range(1, len(xData)+1), yData)

## - 그래프 기본 요소 설정
plt.title('기본 막대그래프')
plt.xlabel('X 데이터')
plt.ylabel('Y 데이터')

## - 화면에 그래프 출력
plt.show()
```

```python
## ===================================================
## [2] 기본 막대 그래프 + DataFrame
## ===================================================

## - 데이터
dataDF = pd.DataFrame( {'A': [10, 50, 30, 40, 20],
                        'B': [15, 35, 45, 25, 55]})
dataDF = dataDF.sort_values(by=['A'], ascending=False).reset_index(drop=True)

## - 막대 그래프 인스턴스 생성 : 기본값으로 선 그래프 생성
##   align 매개변수 : 막대와 라벨 위치 설정
##   width 매개변수 : 막대 너비 설정 [기] 0.8
##   color 매개변수 : 막대 색상 설정

colors=['orange','yellow','green','lightgreen','blue']

# 색상코드값 : '#RRGGBB', '#RRGGBBAA'. 각 자리별 0 ~ F (0 ~ 255)
colors=["#C93030", "#8D9838", "#56E643", "#E13EB3", "#60D6E0"]

bar = plt.bar(dataDF.A.index, dataDF.A, 
              align='edge', width=0.5, color=colors)

## - 그래프 기본 요소 설정
plt.title('기본 막대 그래프')
plt.xlabel('X 데이터')
plt.ylabel('Y 데이터')

## - 화면에 그래프 출력
plt.show()
```

```python
## ===================================================
## [2] 기본 막대 그래프 + DataFrame + 수평막대
## ===================================================

## - 데이터
dataDF = pd.DataFrame( {'A': [10, 50, 30, 40, 20],
                        'B': [15, 35, 45, 25, 55]})
dataDF = dataDF.sort_values(by=['A'], ascending=False).reset_index(drop=True)

## - 막대 그래프 인스턴스 생성 : 기본값으로 선 그래프 생성
##   align 매개변수 : 막대와 라벨 위치 설정
##   width 매개변수 : 막대 너비 설정 [기] 0.8
##   color 매개변수 : 막대 색상 설정

colors=['orange','yellow','green','lightgreen','blue']

# 색상코드값 : '#RRGGBB', '#RRGGBBAA'. 각 자리별 0 ~ F (0 ~ 255)
colors=["#C93030", "#8D9838", "#56E643", "#E13EB3", "#60D6E0"]

bar = plt.barh(dataDF.A.index, dataDF.A, 
              align='center', color=colors)

## - 그래프 기본 요소 설정
plt.title('기본 막대 그래프')
plt.xlabel('X 데이터')
plt.ylabel('Y 데이터')
plt.legend(labels=['A Data'])

## - 화면에 그래프 출력
plt.show()
```

```python
## ===================================================
## [2] 기본 막대 그래프
## ===================================================
## - 데이터
dataDF = pd.DataFrame( {'A': [10, 50, 30, 40, 20],
                        'B': [15, 35, 45, 25, 55]})
## - 컬럼별 합계
sumSR = dataDF.sum()
display(sumSR)

## - 막대 그래프 인스턴스 생성 : 기본값으로 선 그래프 생성
colors=["#C30D9F", "#42DBE0"]

## - 막대그래프 인스턴스 생성
bar = plt.bar(sumSR.index, sumSR.values, color = colors, align='edge')

## - 그래프 기본 요소 설정
plt.title('컬럼별 합계')
plt.xlabel('컬럼')
plt.ylabel('합계')

## - 화면에 그래프 출력
plt.show()
```

```python
## ===================================================
## [3] 응용 막대 그래프
## ===================================================
## - 데이터
dataDF = pd.DataFrame( {'A': [10, 50, 30, 40, 20],
                        'B': [15, 35, 45, 25, 55]})
## - 컬럼별 합계
sumSR = dataDF.sum()
display(sumSR)

## - 막대 그래프 인스턴스 생성 : 기본값으로 선 그래프 생성
colors=["#C30D9F", "#42DBE0"]

## - 막대그래프 인스턴스 생성
bar = plt.bar(sumSR.index, sumSR.values, color = colors, align='edge')

## - 막대그래프 안에 막대별 인스턴스 추출
for rect in bar:
    print(rect, rect.get_height(), rect.get_width(), rect.get_x())
    H, W, X = rect.get_height(), rect.get_width(), rect.get_x()

    plt.text(X+(W/2), H/2, f'합계 : {H}',
             fontdict={'fontsize':15},
             ha='center', va='center')      ## 텍스트 정렬
    
plt.show()
```

```python
## ===================================================
## [3] 응용 막대 그래프 - 수평?
## ===================================================
## - 데이터
dataDF = pd.DataFrame( {'A': [10, 50, 30, 40, 20],
                        'B': [15, 35, 45, 25, 55]})
## - 컬럼별 합계
sumSR = dataDF.sum()
display(sumSR)

## - 막대 그래프 인스턴스 생성 : 수평막대
colors=["#C30D9F", "#42DBE0"]
barh = plt.barh(sumSR.index, sumSR.values, color = colors, align='edge')

## ha : 수평 정렬 설정 ('center', 'left', 'right')
## va : 수직 정렬 설정 ('center','top','bottom')
for rect in barh:
    h = rect.get_height()
    w = rect.get_width()
    x = rect.get_x()
    y = rect.get_y()
    print(f'h => {h}, w=> {w}, x => {x}, y => {y}')
    plt.text(x+w/2.0, y+h/2.0, f'{w}', ha='center',va='center',size=12)

## - 그래프 기본 요소 설정
plt.title("컬럼별 합계")
plt.xlabel("컬럼")
plt.ylabel("합계")

## - 화면에 그래프 출력
plt.show()
```

```python
## ===================================================
## [3] 응용 막대 그래프 - 여러개 데이터를 비교 그래프
## ===================================================
# 데이터 준비
xData  = pd.Series(range(1,8,3))   # 1,4,7
yData1 = pd.Series([50,40,90])
yData2 = pd.Series([80,90,60])
yData3 = pd.Series([20,50,30])

## 시각화 준비
W, W2 = 0.8, 0.8*2

plt.bar(xData,    yData1, width=W)      # xData => [  1,   4,   7]
plt.bar(xData+W,  yData2, width=W)      # xData => [1.8, 4.8, 7.8]
plt.bar(xData+W2, yData3, width=W)      # xData => [2.6, 5.6, 8.6]

plt.xticks(xData+ W, ['1970','1980','1990'])
plt.show()
```
