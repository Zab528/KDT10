### [ 데이터 시각화 - 공통사항 ]
- 모든 그래프에 공통적으로 들어가는 요소들
- 그래프 제목, x축 라벨, y축 라벨
- 한 axes에 여러 개 그래프가 존재하는 경우 식별용 범례(legend)
- x,y축에 눈금(ticks) 2종류 즉, major ticks / minor ticks

```python
### [1] 모듈 로딩
import matplotlib.pyplot as plt
import koreanize_matplotlib
```

```python
## [2] 기본 선그래프
## - 선 그래프 인스턴스 생성 : 기본값으로 선 그래프 생성
##          plot(x, y)       : x축 데이터, y축 데이터
line = plt.plot(['a', 'b', 'c', 'd', 'e'], [10,20,30,40,50])

## - 그래프 공통 요소 설정 => 각 요소별 인스턴스 반환
fontdict  = {'size':20, 'color':'red', 'weight':'bold'}
labeldict = {'size' : 17, 'color' : 'blue'}

## -> 그래프와 타이틀 간격 pad 매개변수 설정
## -> 그래프에 대한 타이틀 위치 loc 매개변수 설정 : 'left', 'center', 'right'
plt.title("[선 그래프]", loc='left', pad=20, **fontdict)

## -> 눈금과 라벨들의 간격 조절 labelpad 매개변수 설정
## -> 라벨들의 위치 loc 매개변수 설정 : x, y라벨에 따라 값 다름
plt.xlabel("종류", labeldict, loc='right', **labeldict)   ## 'left', 'center', 'right'
plt.ylabel("값",   labeldict, loc='top', **labeldict)     ## 'top', 'center', 'bottom'        

## - 화면에 그래프 출력
plt.show()
```

```python
## [3] 사용자 설정 선 그래프
## - 선 그래프 인스턴스 생성 : 기본값으로 선 그래프 생성
##          plot(x, y, fmt)       : x축 데이터, y축 데이터,
##                                  fmt 선스타일설정 => 색상 마커스타일 선스타일
## 여러개 그래프 식별을 위한 범례 설정
## -> 방법1: 그래프에 label속성 설정 + plt.legend()

## label 매개변수 설정 => 범례로 사용
line1 = plt.plot(['a', 'b', 'c', 'd', 'e'], [10, 20, 30, 40, 50], '+r', label='Line1')
line2 = plt.plot(['a', 'b', 'c', 'd', 'e'], [15, 25, 35, 45, 55], 'og', label='Line2')

plt.legend()

## - 화면에 그래프 출력
plt.show()
```

```python
## [3] 사용자 설정 선 그래프
## - 선 그래프 인스턴스 생성 : 기본값으로 선 그래프 생성
##          plot(x, y, fmt)       : x축 데이터, y축 데이터,
##                                  fmt 선스타일설정 => 색상 마커스타일 선스타일
## 여러개 그래프 식별을 위한 범례 설정
## -> 방법2: plt.legend(labels=[])

line1 = plt.plot(['a', 'b', 'c', 'd', 'e'], [10, 20, 30, 40, 50], '+r')
line2 = plt.plot(['a', 'b', 'c', 'd', 'e'], [15, 25, 35, 45, 55], 'og')

## 범례 화면 출력. labels 매개변수 : 그래프 수 만큼 리스트 담아 설정
plt.legend(labels=['L1','L2'])

## - 화면에 그래프 출력
plt.show()
```

```python
## [4] 사용자 설정 선 그래프
## - 선 그래프 인스턴스 생성 : 기본값으로 선 그래프 생성
##          plot(x, y, fmt, **kwargs)       : x축 데이터, y축 데이터,
##                                            fmt 선스타일설정 => 색상 마커스타일 선스타일
##                                            **kwargs       => Line2D의 속성들 설정
line1 = plt.plot(['a', 'b', 'c', 'd', 'e'], [10, 20, 30, 40, 50], '+r', markersize=20)
line2 = plt.plot(['a', 'b', 'c', 'd', 'e'], [15, 25, 35, 45, 55], 'og', markeredgecolor='yellow')

## - 화면에 그래프 출력
plt.show()
```

```python
## [5] 사용자 설정 선 그래프
## - 선 그래프 인스턴스 생성 : 기본값으로 선 그래프 생성
##          plot(x, y, fmt, **kwargs)       : x축 데이터, y축 데이터,
##                                            fmt 선스타일설정 => 색상 마커스타일 선스타일
##                                            **kwargs       => Line2D의 속성들 설정
## Line2D 속성값들 저장 => Line2D 속성명:값
propertys = {'ms':20,
             'mec':'yellow',
             'ls':'-.'}
line1 = plt.plot(['a', 'b', 'c', 'd', 'e'], [10, 20, 30, 40, 50], '+r', markersize=20)
line2 = plt.plot(['a', 'b', 'c', 'd', 'e'], [15, 25, 35, 45, 55], 'og', **propertys)

## - 화면에 그래프 출력
plt.show()
```

```python
## [6] 그래프의 인스턴스 속성 읽기 : 객체변수명.get_속성명()
line1_, line2_ = line1[0], line2[0]

print(f'Line1 그래프 속성들 -----------\n')
print(f'마커사이즈 : {line1_.get_ms()}')
print(f'마커색  상 : {line1_.get_mfc()}')

print(f'Line2 그래프 속성들 -----------\n')
print(f'마커사이즈 : {line2_.get_ms()}')
print(f'마커색  상 : {line2_.get_mfc()}')
print(f'마커테두리 : {line2_.get_mec()}')
print(f'선  스타일 : {line2_.get_ls()}')
print(f'데  이  터 : {line2_.get_data()}')
```
